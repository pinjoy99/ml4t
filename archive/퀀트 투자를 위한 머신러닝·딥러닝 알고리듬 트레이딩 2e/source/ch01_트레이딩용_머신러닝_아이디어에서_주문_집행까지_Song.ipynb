# Intro

- 알고리즘 거래 및 머신 러닝 소개:
    - 알고리즘 거래: 컴퓨터 프로그램을 사용하여 거래 전략을 자동화하는 개념.
    - 머신 러닝: 데이터에서 규칙이나 패턴을 학습하여 목표를 달성하는 알고리즘.
- 투자 활동에 머신 러닝의 적용:
    - 시장 관찰과 데이터 분석을 통한 미래 기대 형성 및 거래 결정.
    - 포트폴리오 관리를 통한 수익과 리스크의 관리.
- 알파 생성을 위한 활동:
    - 알파: 벤치마크를 상회하는 포트폴리오 수익.
    - 능동적 관리의 핵심 법칙: 정보 비율 (IR)을 통한 알파 생성.
    - 정보 비율 (IR) 구성 요소: 정보 계수 (IC)와 전략 범위의 제곱근.
- 머신 러닝의 중요성:
    - 경쟁적 시장에서 정확한 예측의 필요성.
    - 머신 러닝의 역할: 다양한 데이터를 활용하여 더 나은 예측 및 투자 결정 제공.
- 알고리즘 거래의 확장:
    - 과거의 알고리즘 거래: 거래 실행 자동화에 초점.
    - 현재의 접근 방식: 아이디어 생성부터 전략 평가까지의 포괄적 관점.
- 산업 트렌드와 머신 러닝:
    - 머신 러닝의 경쟁 우위로 인한 투자 산업 변화.
    - 머신 러닝의 투자 프로세스 적용 방법.

# 투자업계에서 머신러닝의 부상

- 투자 산업의 변화:
    - 지속적인 경쟁 증대, 기술적 발전, 어려운 경제 환경 속에서 투자 산업이 급격한 변화를 겪고 있음.

- 알고리즘 거래와 머신 러닝의 부상을 촉발시킨 주요 트렌드:
   - 전자 거래의 확산과 시장 통합
   - 자산군과 반대되는 리스크 팩터 노출에 기반한 투자 전략의 발전
   - 컴퓨팅 파워, 데이터 관리, 통계적 방법 등에서의 혁신
   - (인간의 재량적 투자 대비) 알고리즘 거래의 우수한 성과 

- 금융 위기의 영향:
   - 2001년과 2008년의 금융 위기가 다양화와 리스크 관리에 접근하는 방식에 영향을 미침.
   - 낮은 수익률과 낮은 변동성으로 인한 패시브 투자 수단인 ETF의 부상.

- 경쟁 압력의 반영:
   - 헤지 펀드 수수료의 하락으로 나타남.
   - 연간 관리 수수료와 이익의 변화.

## 전자 거래에서 고빈도 매매(HFT)까지

- **전자 거래의 발전:**
    - 1960년대 이후 전자 거래는 네트워크를 통해 가격을 라우팅하고 기능, 거래량, 자산군의 범위가 극적으로 발전했다.
    - 이러한 발전은 주식시장을 주도하며, 해리스와 스트루메이어에 의해 포괄적으로 다루어졌다.
- **ECN과 다크풀:**
    - SEC의 주문 처리 규칙은 ECN을 도입하여 주식 거래소 경쟁을 촉진했다.
    - 다크풀은 주문장을 관리하고 대량 주문을 비공개로 처리하여 거래소와 경쟁한다.
- **알고리듬 거래와 DMA:**
    - 비용 효율적인 주문 집행 알고리듬은 거래량과 자산군에 걸쳐 널리 사용되었다.
    - DMA는 거래자에게 더 강력한 실행 제어를 제공하여 직접 주문을 거래소로 전송한다.
- **고빈도 매매와 규제:**
    - 고빈도 매매는 미세한 시간 간격으로 자동화된 거래를 통해 작은 수익을 얻는 전략이다.
    - 고빈도 매매는 시장의 취약성과 변동성에 대한 우려를 증가시켰으며, 규제 기관의 관심을 끌었다.
- **산업 통합과 미래 전망:**
    - 고빈도 매매의 수익은 감소했지만 산업은 공유된 인프라 투자와 산업 통합을 통해 변화하고 있다.
    - 스타트업은 고빈도 매매를 민주화하는 새로운 기회를 제공하고 있다.

## 팩터 투자와 스마트 베타 펀드

- **자산 수익률과 리스크 팩터:**
    - 자산 수익률은 금융 투자의 불확실성이나 리스크에 의해 결정된다.
    - 예를 들어, 주식 투자는 기업의 사업 리스크를 내포하고 있고, 채권 투자는 부도 리스크가 있다.
    - 머신러닝을 사용하여 이러한 리스크 팩터의 행태를 식별하고 예측하여 투자 전략을 설계하는 것이 중요하다.
- **모던 포트폴리오 이론과 CAPM:**
    - 모던 포트폴리오 이론(MPT)은 자산의 고유 리스크와 체계적 리스크를 구별한다.
    - CAPM은 시장 포트폴리오에 대한 수익률을 설명하기 위한 모델로, 시장 베타와 무위험 이자율을 고려한다.
- **리스크 프리미엄과 추가 리스크 팩터:**
    - CAPM의 예측에 의문을 제기하며, 수많은 추가 리스크 팩터가 발견되었다.
    - 이러한 팩터는 주식 시장의 성과를 설명하고, 가치, 규모, 모멘텀 효과 등을 포함한다.
- **멀티팩터 모델과 스마트 베타 펀드:**
    - 멀티팩터 모델은 시장 포트폴리오보다 더 넓은 리스크로 정의되며, APT는 분산화할 수 없는 여러 체계적인 리스크 팩터로 보상을 받는다고 주장한다.
    - 스마트 베타 펀드는 패시브 전략을 취하지만 더 나은 수익률을 위해 팩터를 조정하는 하이브리드 투자 전략이다.
- **리스크 팩터와 머신러닝:**
    - 리스크 팩터의 발견과 예측은 자산 수익률에 중요한 영향을 미치며, 투자 업계에서 머신러닝 적용이 급증하는 이유 중 하나이다.

## 알고리즘 개척자는 인간보다 우위에 있다

- 알고리즘 거래 선구자들의 성과와 성장:
   - 알고리즘 거래를 주도한 기업들의 자산 운용 규모와 성적표는 투자자들의 관심을 유발하고, 이에 따른 산업 노력들이 그들의 성공을 복제하려는 역할을 하였다.
- 시스템적 전략의 도입:
   - 주로 또는 전적으로 알고리즘 의사 결정에 의존하는 체계적인 전략은 1982년 수학자 제임스 시몬스가 창립한 르네상스 테크놀로지스를 통해 가장 유명하게 소개되었다.
   - 르네상스 테크놀로지스의 비밀스러운 메딜리온 펀드는 외부인에게 닫혀있지만, 1982년 이래 약 35%의 연평균 수익률을 기록했다.
- 양적 헤지 펀드들의 부상:
   - 알고리즘을 기반으로 한 체계적 전략을 사용하는 양적 헤지 펀드 중 가장 유명한 DE Shaw, Citadel, Two Sigma 등은 2017년 처음으로 투자자들에게 수수료를 고려한 총 수익 상위 20위에 올랐다.

**머신러닝 기반 펀드 1조 달러 운용 자산 유치**

- 알고리즘 전략은 최근 6년간 연평균 15% 성장하여 헤지 펀드, 상장지수펀드, 스마트 베타 ETF 등 약 1.5조 달러를 관리하고 있다.
- 양적 헤지 펀드 산업은 2010년 이후 거의 두 배로 성장하여 약 1조 달러의 자산 운용 규모에 도달했다.
- 전체 헤지 펀드 산업 자본은 최신 보고서에 따르면 3.21조 달러에 달한다.

**퀀트멘달 펀드의 출연**

- 활동적 투자 관리에서는 두 가지 접근 방식이 있다: 체계적(퀀트)과 재량적 투자.
    - 체계적(퀀트) 투자는 알고리즘을 사용하여 데이터 기반으로 투자 기회를 식별하는 방식. 반복 가능하고 정량적인 모델을 사용하여 투자 결정을 내리는 것이 특징.
    - 재량적 투자는 투자자의 주관적인 판단과 경험을 기반으로 적은 수의 증권에 대해 심층적으로 분석하고 투자하는 방식. 주로 투자자의 직관과 판단이 투자 결정에 영향을 미치는 특징.
- 기본적인 트레이더들도 퀀타티브 기법을 사용하며, 이는 다양한 증권에 대한 패턴과 동적을 거래하는 퀀트 펀드로 이어진다.
- 퀀트 펀드는 전체 헤지 펀드 자산의 약 17%를 차지하고 있다.

**전략적 능력에 대한 투자**
- 다음 세가지 트랜드로 인해 알고리즘 트레이딩 전략에서 데이터 사용이 증가 하였다
	- 디지털 데이터 가용성의 기하급수적 증가
	- 낮은 비용의 컴퓨팅 성능 및 데이터 저장 용량 증가
	- 복잡한 데이터 세트를 분석하기 위한 통계적 방법의 발전 

## 머신러닝과 대체 데이터

- 헤지 펀드의 알파 탐색:
    - 헤지 펀드는 오랫동안 정보적 우위와 새로운 상관관계가 없는 신호를 발견하는 능력을 통해 알파를 찾았습니다.
    - 과거에는 쇼핑객이나 선거나 공청회 등의 자체 조사를 포함했습니다.
    - 그러나 2010년 이후, 내부자 정보를 사용한 거래로 인한 트레이더, 포트폴리오 매니저 및 분석가들의 일련의 기소가 업계에 충격을 줌
- ML을 통한 정보적 우위:
    - 기존 및 대체 데이터 소스를 활용한 정보적 우위는 전문가 및 산업 네트워크나 기업 경영진에 대한 접근과는 관련이 없습니다.
    - 대신, ML을 사용하여 대량의 데이터를 수집하고 실시간으로 분석하는 능력에 기인합니다.    

## 크라우드 소싱 거래 알고리즘

- 크라우드 소싱 플랫폼 소개:
    - 일부 알고리듬 트레이딩 회사들이 크라우드 소싱 리스크 팩터에 데이터와 프로그래밍 환경을 제공하는 투자 플랫폼을 제공하기 시작했다.
    - 대표적인 사례로는 월드퀀트, 권토피안, 알파 트레이딩 랩스가 있다.
- 월드퀀트의 운영:
    - 2007년에 밀레니엄 자산 관리에서 분리된 월드퀀트는 약 50억 달러를 관리하고 있다.
    - 퀀트 조립 라인으로 구성된 투자 프로세스를 운영하며, 세계 수백 명의 과학자와 시간제 노동자를 고용하고 있다.
- 알파 트레이딩 및 알고리듬:
    - 월드퀀트는 성공적으로 테스트된 4백만 개의 알파 팩터를 생산하고 있으며, 이를 통해 알고리듬을 구성한다.
    - 각 알파 팩터는 미래 자산 가격 변화를 예측하는데 사용되는 알고리듬이다.
    - 다양한 팀은 이러한 알파 팩터를 결합하여 전략을 구성하고, 이를 포트폴리오에 통합하여 펀드를 할당하며, 리스크를 관리한다.

## 데이터의 소싱과 관리

- **데이터 가용성의 진화와 머신러닝의 적용**:
   - 데이터 가용성의 증가는 머신러닝을 트레이딩에 보완적으로 적용하는 핵심 요소가 됨.
   - 새로운 데이터 소스 획득을 위한 업계 지출 증가.
- **데이터 소스 식별과 평가**:
   - 시장, 펀드멘탈, 대체 데이터 등 다양한 소스를 평가하고 식별.
   - 알파 신호를 포함한 안정적인 데이터 소스 추적.
- **데이터 인프라 및 분석 도구 활용**:
   - 클라우드 기반의 확장 가능한 데이터 인프라 구축.
   - 하둡 또는 스파크와 같은 분석 도구를 활용하여 빠르고 유연한 데이터 액세스 가능하게 함.
- **데이터 관리 및 선별**:
   - PIT Point-n-Time 기준으로 데이터를 조정하여 선견자 편향을 피함.
   - 왜곡된 과거 데이터를 제거하고 실제 거래에 적합한 데이터로 훈련.
- **머신러닝 알고리즘의 훈련과 관리**:
   - 왜곡된 과거 데이터로 훈련된 알고리즘은 실제 거래 시 손실을 야기할 수 있음.
   - 실제 거래에 적합한 데이터로 머신러닝 알고리즘을 신중히 관리하고 훈련해야 함.

## 알파 팩터 리서치에서 포트폴리오 관리까지

- 알파 팩터는 데이터에서 신호를 추출하여 특정 투자 영역의 수익을 예측하는데 사용된다. 
- 이는 머신러닝 워크플로에서 도메인 영역별 특성으로 볼 수 있다. 
- 알파 팩터를 활용하기 위해서는 리서치 스텝과 실행 스텝이 필요하다.

**리서리 스텝**

- **리서치 스텝**:
   - 알파 팩터 설계와 평가가 포함됨.
   - 예측 팩터는 데이터 소스와 중요한 입력 변수 간의 체계적인 관계를 포착함.
   - 예측력 최적화를 위해 창의적인 특성 공학이 필요함.
- **데이터 마이닝의 주요 리스크**:
   - 머신러닝으로 인한 잘못된 발견은 신중한 관리가 필요한 주요 리스크
   - 리스크 감소를 위해 검색 프로세스에 집중하는 것이 중요.
   - 투자자들은 이론과 일치하는 팩터를 선호하지만 실제로는 실용적인 측면을 고려해야 함.
- **알파 팩터의 신호 검증**:
   - 대표적인 콘텍스트에서 예측력에 대한 강력한 추정 필요.
   - 신뢰할 수 있는 추정치를 훼손하는 방법론적, 실용적인 함정 존재.
   - 데이터의 생존자 편향이나 선견자 편향을 고려하여 신뢰성 있는 결과를 얻어야 함.
   - 4장에서 이러한 프로세스를 관리하는 방법을 다룸.

**실행 스텝**

- **알파 팩터의 실행 스텝**:
   - 알파 팩터는 매수 또는 매도 주문으로 이어지는 신호를 발산함.
   - 포트폴리오 보유는 전체 포트폴리오 리스크와 상호작용하며 특정 리스크 프로파일에 영향을 줌.
- **포트폴리오 관리**:
   - 투자 목표와 일치하는 포트폴리오 수익과 리스크 균형을 맞추기 위해 포지션 크기를 최적화함.
- **포트폴리오 최적화와 거래 전략 워크플로**:
   - 5장에서는 포트폴리오 최적화와 거래 전략 워크플로에 적용할 수 있는 주요 기술과 도구를 다룸.

## 전략 백테스팅
- **과학적 접근**:
   - 알고리듬 전략에 투자 아이디어 통합을 위해 과학적 접근 필요
   - 시장이 다양한 조건에서 변화하더라도 전략이 여전히 유효한지 확인하는 것이 중요
- **검증 및 테스팅**:
   - 검증에는 대안적 시나리오를 포착하기 위한 시뮬레이션 데이터 사용.
   - 백테스팅 엔진으로 편향 없는 성과 추정치를 얻고 현실적인 실행을 시뮬레이션 함.
   - 백테스팅 엔진은 이러한 데이터 유발 편향과 결함을 고려하여 실제 시장 상황을 정확하게 모사하고, 
   - 실제 거래 과정에서 발생할 수 있는 다양한 상황을 반영할 필요가 있다
- **백트레이딩과 집라인 활용**:
   - 8장에서는 백트레이딩과 집라인을 사용하는 방법을 다룸.
   - MLAT 워크플로의 소개를 완료하면서 다양한 방법론적 과제를 탐색함.

# 트레이딩을 위한 머신러닝 : 전략과 사용사례

## 알고리즘 트레이딩의 진화

- **퀀트 전략의 진화**:
   - 1980년대와 1990년대: 학술적 연구에서 나온 신호를 활용한 퀀트 전략.
   - 2000년대: 팩터 기반의 투자가 확산됨.
   - 현재: 머신러닝과 대체 데이터를 활용한 반복 가능한 거래 전략이 주목받음.  
- **주요 목표와 거래 전략**:
   - 우호적인 가격 결정을 목표로 하는 거래 실행 알고리듬.
   - 단기 거래를 통한 이익 추구.
   - 다른 시장 참여자의 행동 예측을 목표로 하는 행태 거래 전략.
- **거래 실행 프로그램**:
   - 시장 영향을 제한하는 거래 실행 프로그램의 다양한 유형.
   - 시간 가중 평균 가격(TWAP), 거래량 가중 평균 가격(WAP) 등 다양한 거래 전략의 활용.
- **고빈도 매매 펀드(HFT fund)**:
   - 매수-매도 스프레드 기반, 작은 가격 움직임에서 이익을 추구하는 펀드.
   - 매우 짧은 시간 보유
   - 행태 주의 알고리듬을 활용한 거래 전략은 저우동성 환경에서 운영
   - 스니핑 알고리즘 : 다른 시장 참여자들의 전략에 대한 통찰력 생성(더 큰 플레이어들의 움직임 예상)

## 거래를 위한 머신러닝 사용 사례

- **머신러닝의 다양한 활용**:
   - 시장, 펀더멘털, 대체 데이터에서 신호 추출 가능.
   - 계량화 가능한 목표와 디지털 데이터를 활용하여 의사결정을 지원하고 자동화하는 유연한 도구.
- **머신러닝의 적용 사례**:
   - 데이터 마이닝을 통한 패턴 확인과 특성 추출.
   - 리스크 팩터나 알파 생성을 위한 지도학습.
   - 개별 신호를 전략으로 통합하여 거래 아이디어 생성.
   - 알고리듬에 의한 리스크 프로파일에 기반한 자산 배분.
   - 합성 데이터를 사용한 전략의 테스트와 평가.
   - 강화학습을 활용한 자동화된 대화형 전략 개선.

**특성 추출과 통찰력을 위한 데이터 마이닝**

- **정보 이론을 활용한 평가**:
   - 후보 특성의 신호 내용을 평가하기 위해 정보 이론을 사용.
   - 자산 수익률 예측을 위한 지도학습 알고리듬에서 상호 정보량을 활용.
- **비지도학습을 통한 구조 식별**:
   - 클러스터링과 차원 축소를 활용하여 데이터의 구조를 식별하는 방법을 소개.
   - 금융 텍스트 데이터를 요약하기 위해 베이지안 확률 모델 적용.
   - 딥러닝을 사용하여 자산 특성에 따라 조건화된 비선형 리스크 팩터 추출.
- **모델 투명성 강조**:
   - 모델의 예측력에 대한 통찰력을 얻기 위해 SHAP(SHapley Additive exPlanations) 소개.
   - 그래디언트 부스팅 머신에 모델 투명성을 적용하는 방법을 설명.

**알파 팩터 창출을 위한 지도학습**

- 머신러닝을 거래에 적용하는 주된 이유는 자산의 펀더멘털, 가격 변동, 시장 조건 등을 예측하기 위함
- 이를 위해 다양한 머신러닝 알고리즘을 활용
    - **다운스트림 모델**: 
        - 이는 개별 자산의 전망, 자본 시장 기대, 증권 간의 상관관계 등을 예측하여 포트폴리오 수준에서 신호를 생성하는 것을 의미
    - **예측의 영향**: 
        - 머신러닝 예측은 퀀트멘탈 접근법과 임의의 거래에 영향을 줄 수 있습니다. 특정 리스크 요소나 기술적 접근 구현을 타깃으로 할 수 있다
    - **입력 변수 생성**: 
        - 3장에서는 머신러닝 기반 평가 모델을 위한 입력 변수를 생성하는 방법에 대해 다룬다
    - **대체 데이터 활용**: 
        - 14장, 15장, 16장에서는 대체 데이터를 사용하여 프로젝트 수익을 평가하는 방법을 소개
    - **시장 기대 예측**: 
        - 9장에서는 시장 기대를 위한 입력 변수로 매크로 변수를 예측하는 방법과 변동성 등의 리스크 요소 소개
    - **순환 신경망 소개**: 
        - 19장에서는 비선형 시계열 데이터에서 우수한 성능을 달성하는 순환 신경망에 대해 소개

**자산 배분**

- 의사결정 트리 모델을 활용하여 머신러닝은 리스크 패리티의 계층적 형태를 계산하고 포트폴리오를 구성하는 데 사용되었습니다.
- 리스크 패리티는 포트폴리오 내 각 자산이 전체 리스크에 동일하게 기여하도록 보장하는 투자 전략이다.
- 자산 가격 패턴이 리스크 특성을 주도하고, 이를 통해 우수한 리스크 리턴 특성을 달성할 수 있습니다.
- 계층적 클러스터링을 통해 보다 나은 상관관계 패턴을 반영하는 데이터 기반 리스크 클래스를 추출합니다.

**거래 아이디어 테스트**

- **백테스팅(Backtesting):**
    - 성공적인 알고리즘 트레이딩 전략 선택의 중요한 단계. 
    - 교차 검증을 포함한 적절한 방법과 결합하여 신뢰할 수 있는 표본 외 결과를 얻는 머신러닝 기법이다.
- **교차 검증(Cross-validation):**
    - 합성 데이터를 활용하여 다중 테스트를 수정하기 위한 핵심적인 머신러닝 기법. 
    - 표본 외 오차의 건전한 추정치를 얻기 위해 시장, 기본, 대체 데이터를 활용한 머신러닝 모델을 테스트하는 다양한 방법을 시연한다.
- **합성 데이터 활용:**
    - 금융 데이터의 시계열 특성은 선견 편향을 피하기 위해 표준적인 접근법으로 수정할 필요가 있다. 
    - 과거 데이터의 제한된 가용성으로 인해 합성 데이터를 사용하는 대안적인 접근법이 발전하고 있다.
- **적대적 생성 네트워크(GANs):**
    - 고품질 합성 데이터를 생성하는 데 사용되는 기법으로, 21장에서 소개된다.

**강화 학습**

- **강화학습(Reinforcement Learning):**
    - 시장에서 경쟁적이고 상호작용하는 환경에서 에이전트를 훈련하는 머신러닝 기법.
    - 보상을 기반으로 하여 정책 함수를 학습시키는 것이 목적.
    - 금융 머신러닝에서 주목받는 영역 중 하나로 간주되고 있음.
    - Hendricks and Wilcox (2014) 및 Nermyvaka, Feng and Kearns (2006)의 거래 실행 애플리케이션을 참고할 수 있음.
- **OpenAI Gym을 활용한 거래 강화학습:**
    - 22장에서 OpenAI Gym을 사용하여 거래를 위한 강화학습 알고리듬을 훈련하는 방법을 소개.
    - Q-learning과 같은 핵심 강화 알고리듬을 제시한다.
