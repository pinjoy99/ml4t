### 대체 데이터의 원천

대체 데이터 속성이 새로운 데이터 원천으로 발전함에 따라 이전의 레이블화 데이터는 주류의 일부분이 됨<br>
발틱 운임 지수 → 블룸버그 터미널

발틱운임지수: 석탄, 철광석, 시멘트, 곡물 등 원자재를 싣고 26개 주요 해상운송경로를 지나는 선적량 15천 톤 이상 선박의 화물운임과 용선료 등을 종합해 산정하는 지수<br>
배들이 원자재를 '얼마나 많이' 싣고, '얼마나 자주' 돌아다니는지를 알려주는 지표

데체 데이터에 가공 데이터가 포함. 감성 점수과 같은 거래 가능한 신호를 추출 목표

대체 데이터 소스가 알고리즘 트레이딩에 대한 가치나 신호 내용을 결정

개인 데이터: SNS, 제품의 관심이나 인식 등의 전자 상거래 활동, 검색 엔진 활동, 스마트폰 앱, 다운로드, 메시지 전송 내역 등...<br>
예. 소셜미디어 감성 분석 → 개별 주식, 산업 바스켓, 시장 지수에 적용 가능

### 대체 데이터 평가를 위한 기준

대체 데이터의 궁극적인 목표는 양이며 상관관계가 없는 투자 수익율(알파)을 창출하는 트레이딩 신호에 경쟁력 있는 정보 제공<br>
결국 대체 데이터를 사용하여 트레이딩 신호 추출<br>
예. Sharp 비율이 높게 만드는 정보

샤프 비율: <br>
- 사전적: (기대되는 투자의 수익률 - 벤치마크 수익률)의 기대값을 (기대되는 투자의 수익률 - 벤치마크 수익률)의 표준편차로 나눈 것
- 사후적: 기대값이 아닌 (실현된 수익률 - 벤치마크 수익률)의 평균을 사용<br>
-
예를 들어 1년동안 코스피가 20% 오르고 내 포트폴리오가 30% 올랐는데, 내 포트폴리오의 월간 수익률에서 코스피의 월간 수익률을 뺀 값의 표준편차가 5%라면, 사후적 샤프 비율은 (0.3-0.2)/(0.05*12^(1/2)) = 약 0.57<br>
높은 샤프지수의 투자일수록 더 적은 위험으로 더 높은 수익률을 내는 것<br>
따라서 높을 수록 더 좋은 포트폴리오<br>
단, 수익률의 분포가 정규분포와 크게 차이가 나는 투자대상이거나, 꾸준히 작은 수익을 내다가 한번 크게 잃는 투자전략 같은 경우에는 샤프지수로 판단하는 것이 적절하지 않음

퀀트사는 개별적인 추출 신호 보다는 여러 추출 신호를 결합하여 수익을 만들 수 있는 알파 팩터 라이브러리 구축 중

경쟁으로 인한 신호 예측력 감소 문제...<br>
대체 데이터 세트에서 신호 내용의 반감기를 연장하기 위한 전략: 독점 계약이나 진입 장벽을 높이기 위한 처리 문제를 제기하는 데이터 세트에 대해 초점

### 대체 데이터로 작업

웹 스크래핑을 사용한 대체 데이터 수집 실습<br>
오픈테이블 식당 데이터 + Seeking Alpha의 Earning call transcript

대체 데이터의 일반적인 소스는 직원 의견, 고객 리뷰를 통해 인사이트를 얻을 수 잇는 리뷰 웹 사이트(글래스도어, 옐프)<br>
기업 전망이나 시장 가치를 예측하거나 거래 신호를 얻기 위한 머신러닝 모델의 입료자료로 효용성<br>
단, HTML 소스에서 추출, 법적 장애물X

웹 데이터 가져오기 위한 절차
1. 타겟 웹사이트 조사: 크롤링하려는 웹사이트의 구조와 URL 패턴을 파악
2. Request: 웹사이트의 데이터를 가져오기 위해 해당 웹사이트에 요청
3. HTML 응답 파싱: 받아온 HTML 응답을 파싱하여 필요한 데이터를 추출
4. 데이터 저장: 추출한 데이터를 원하는 형식으로 저장 (CSV, JSON 등)
5. 반복: 모든 원하는 데이터를 크롤링할 때까지 2-4단계를 반복

Requests: 웹사이트 접근하여 데이터를 가져오기 위한 라이브러리<br>
BeautifulSoup: HTML 및 XML 문서를 파싱하고 원하는 데이터를 추출하기 위한 Python 라이브러리
정적 데이터만 처리(JavaScript로 생성되는 동적인 데이터나 내용을 처리할 수 없음)<br>
로그인, 스크롤 다운, 버튼 클릭 등의 웹 상호작용 기능 없음

Selenium: 웹 브라우저를 자동화하기 위한 도구, JavaScript로 렌더링된 페이지나 사용자 상호작용을 모방할 때 유용<br>
동적 데이터 처리, JavaScript로 로드되는 동적인 웹페이지 내용도 가지고 옴<br>
로그인, 스크롤, 버튼 클릭, 폼 제출 등의 상호작용을 자동화하여 수행<br>
다양한 웹 브라우저 지원<br>
다양한 언어 지원<br>
단점: 웹 브라우저를 띄우고 페이지를 완전히 불러올 때가지 기다리기 때문에 느림<br>
실제 웹 브라우저를 사용하기 때문에 크롤링하는 동안 상대적으로 많은 메모리 소비<br>
웹드라이버 등의 추가적인 설정 필요(웹 드라이버, 브라우저 버전 호환)

```python
import requests		# requests 라이브러리를 import해온다

# 네이버 메인 페이지를 get method를 통해서 받아온다
response = requests.get('https://www.naver.com')
print(response.text)
```

```python
from bs4 import BeautifulSoup as bs

soup = bs(response.text, "html.parser")
presses = soup.select('.thumb_area .thumb') #뉴스를 가져옴
for press in presses:
  print("- Soup 객체: ", press)
  print("- Attrs: ", press.attrs)
  print("- Select One: ", press.select_one("img.news_logo"))
  print("- Attribute: ", press.select_one("img.news_logo").get("src"), press.select_one("img.news_logo").get("alt"))
  print("- Text: ", press.text)
  print()
```
