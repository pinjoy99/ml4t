```python
import requests
import lxml.html
from lxml.cssselect import CSSSelector
import re
```

```python
# 2 베스트셀러 도서정보(도서명, 저자, 가격) 수집
url = 'https://www.hanbit.co.kr/store/books/bestseller_list.html'
params = {}
headers = { 'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36' }

res = requests.get(url, params=params, headers=headers)
# 지정한 url로 부터 html코드를 읽어옴

print(res.encoding, res.headers['content-type'])
print(res.text)
# 서버에서 응답으로 넘어온 html코드 출력
```

```python
html = res.text
root = lxml.html.fromstring(html)
# 응답으로 넘어온 코드를 분석해서 추출하기 쉽게 객체화 함

titles = []
authors = []
bkprices = []

# 추출할 도서정보를 저장하기 위해 리스트 선언

for title in root.cssselect('p.book_tit > a'):
    t = title.text_content()
    print(t)
    titles.append(t)
# 도서명 추출해서 titles 리스트에 저장
```

```python
for writer in root.cssselect('p.book_writer'):
    w = writer.text_content().replace(',','/')
    print(w)
    authors.append(w)
# 저자를 추출해서 authors 리스트에 저장
```

```python
fkey = [',', '원']    # 불용어 목록 정의
for price in root.cssselect('span.price'):
    p = price.text_content()
    p = re.sub(fkey[0], '', p)
    p = re.sub(fkey[1], '', p)
    # 문자열에 포함된 불용어를 찾아서 공백으로 치환
    print(p)
    bkprices.append(p)
# 가격을 추출해서 bkprices 리스트에 저장
```

```python
for i in range(len(titles)):
    print(f'{titles[i]}, {authors[i]}, {bkprices[i]}')
# 각 리스트에 저장된 값들을 하나의 행으로 구성해서 출력

fname1 = './hb_bestseller.csv'
with open(fname1, 'w', encoding='UTF-8') as f:
    for i in range(len(titles)):
        data = f'"{titles[i]}", "{authors[i]}", {bkprices[i]}\n'
        f.write(data)
# 각 리스트에 저장된 값들을 하나의 행으로 구성해서
# hb_bestseller.csv로 저장
```

```python
import pandas as pd

#페이지가 넘어갈때(매 30마다) 2개의 줄이 추가되는 것 확인
종목명 = [] #1
코드 = [] #2
유형 = [] #3
순자산 =[] #4
종가 =[] #5
거래량 = [] #6
수익율 = [] #9
총보수 = [] #15
운용사 = [] #16

with open('ETF.txt','r') as f:
    data = f.readlines()
    
    for i, j in enumerate(data):
        k = i//(30*17)# 페이지수
        m = i-2*k
        if m % 17 == 1:
            종목명.append(j.replace('\n',''))
        elif m % 17 ==2:
            코드.append(j.replace('\n',''))
        elif m % 17 ==3:
            유형.append(j.replace('\n',''))
        elif m % 17 ==4:
            순자산.append(j.replace('\n',''))
        elif m % 17 ==5:
            종가.append(j.replace('\n',''))
        elif m % 17 ==6:
            거래량.append(j.replace('\n',''))
        elif m % 17 ==9:
            수익율.append(j.replace('\n',''))
        elif m % 17 ==15:
            총보수.append(j.replace('\n',''))
        elif m % 17 ==16:
            운용사.append(j.replace('\n',''))

data = {'종목명':종목명,
        '코드':코드,
       '유형':유형,
       '순자산':순자산,
       '종가':종가,
       '거래량':거래량,
       '수익율':수익율,
       '총보수':총보수,
       '운용사':운용사}

# data_m = {i:data[i][:1174] for i in data.keys()}
df = pd.DataFrame(data, columns=['종목명','코드','유형','순자산','종가','거래량','수익율','총보수','운용사'])
```

```python
종목명 = [] # 0
코드 = [] # 1
유형 = [] # 2
순자산 = [] # 3
종가 = [] # 4
거래량 = [] # 5
수익율 = [] # 6
총보수 = [] # 7
운용사 = [] # 8

with open('ETF.txt', 'r') as f:
    data = f.readlines()
    
    for i, j in enumerate(data):
        k = i // (30 * 17)  # 페이지 수
        m = i - 2 * k
        if m % 17 == 0:
            종목명.append(j.replace('\n', ''))
        elif m % 17 == 1:
            코드.append(j.replace('\n', ''))
        elif m % 17 == 2:
            유형.append(j.replace('\n', ''))
        elif m % 17 == 3:
            순자산.append(j.replace('\n', ''))
        elif m % 17 == 4:
            종가.append(j.replace('\n', ''))
        elif m % 17 == 5:
            거래량.append(j.replace('\n', ''))
        elif m % 17 == 6:
            수익율.append(j.replace('\n', ''))
        elif m % 17 == 7:
            총보수.append(j.replace('\n', ''))
        elif m % 17 == 8:
            운용사.append(j.replace('\n', ''))

data = {
    '종목명': 종목명,
    '코드': 코드,
    '유형': 유형,
    '순자산': 순자산,
    '종가': 종가,
    '거래량': 거래량,
    '수익율': 수익율,
    '총보수': 총보수,
    '운용사': 운용사
}

df = pd.DataFrame(data)
```

```python
종목명= df.loc[df['코드'].apply(lambda x : len(x))!=6,'운용사'].values
유형= df.loc[df['코드'].apply(lambda x : len(x))!=6,'종목명'].values
df.loc[df['코드'].apply(lambda x : len(x))!=6,'유형'] = 유형
df.loc[df['코드'].apply(lambda x : len(x))!=6,'종목명'] = 종목명
df.to_csv('./ETF.csv',sep=',',encoding='utf-8')
```
