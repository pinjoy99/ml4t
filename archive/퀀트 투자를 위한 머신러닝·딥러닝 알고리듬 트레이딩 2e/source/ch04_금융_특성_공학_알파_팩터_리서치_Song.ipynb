- 4장의 주요 내용
    - 존재하는 알파 팩터의 범주, 작동 방식, 측정 방법
    - 넘파이, 판다스, TA-Lib를 사용한 알파 팩터 생성
    - 웨이블릿과 칼만 필터를 사용한 데이터 잡음 제거 방법
    - 집라인을 사용한 개별 및 다중 알파 팩터 테스트
    - 정보 계수(Coefficient)를 사용하여 예측 성과 및 회전율을 평가하는 방법

# 알파 팩터 실무 : 데이터 부터 신호까지

1. **알파 팩터의 목표**: 자산 가격의 움직임을 예측하기 위해 기본 데이터를 변환하는 것입니다.
2. **디자인 목적**: 자산의 수익률을 높이면서 리스크를 식별하는 데 목적이 있습니다.
3. **구성**: 하나 또는 여러 입력 변수를 결합하지만, 각 자산에 대해서는 단 하나의 값을 사용합니다.
4. **트레이딩 결정**: 여러 자산에 걸쳐 상대적인 팩터 값이나 단일 자산에 대한 패턴을 기반으로 합니다.
5. **연구 단계 중요성**: 알고리즘 거래 전략을 개발할 때 알파 팩터의 설계, 평가, 조합은 연구 단계에서 중요한 부분입니다.

6. **내용 구성**: 4장에서는 알파 팩터를 연구하는 방법에, 5장에서는 실제로 실행하는 방법에 초점을 맞춥니다. 이후 내용에서는 머신러닝을 활용해 데이터에서 새로운 팩터를 학습하고, 여러 알파 팩터 신호를 효과적으로 결합하는 방법을 다룹니다.

## 알파 팩터 공학

1. **알파 팩터란?** 시장 데이터, 기본적 데이터, 대체 데이터를 분석하여 자산 가격의 움직임을 예측하는 변환된 신호입니다.
2. **팩터의 종류**:
   - **경제적 팩터**: 성장률, 물가 상승률, 변동성 등 거시경제 데이터를 묘사합니다.
   - **투자 스타일 팩터**: 가치-성장 투자, 모멘텀 투자와 같은 특정 투자 스타일을 나타냅니다.
   - **행동적 팩터**: 금융시장 설정, 투자자 행동의 편향 등 가격 움직임을 설명합니다.
3. **팩터의 경제적 의미**:
   - **합리적 경제이론**: 장기적으로 높은 수익을 제공하는 리스크를 보상한다고 봅니다.
   - **행동 경제학**: 편향된 행동으로 인한 리스크 프리미엄을 설명합니다.
4. **팩터 발견의 중요성**: 새로운 수익률 동인을 발견하고 탐색하는 것은 지속적으로 이루어지는 과정입니다. 예를 들어, 250개 이상의 팩터가 이미 확인되었고, 매년 더 많이 발견되고 있습니다.
5. **알파 팩터 생성 방법**: 단순한 수학적 연산, 기본적 분석 등 다양한 방법을 사용합니다.
6. **알파 팩터의 활용**: 과거에는 트레이딩 전략이 특정 기준에 따라 알파 팩터를 활용했습니다. 현대 연구는 다양한 팩터 모델을 개발하여 더 정교한 방법으로 이를 적용하고 있습니다.
7. **머신러닝(ML)의 역할**: 전통적인 공식 대신에 머신러닝은 데이터로부터 직접 신호를 추출하여 알파 팩터를 최적화하는데 효과적입니다. ML 모델은 다양한 팩터를 통합하여 신뢰할 수 있는 예측을 제공할 수 있습니다.

# 수십 년간의 팩터 리서치를 바탕으로 구축

1. **리스크 팩터의 이상적 특성**:
   - **독립성**: 리스크 팩터들은 서로 독립적이어야 합니다.
   - **양의 리스크 프리미엄**: 리스크를 감수하는 대가로 긍정적인 보상을 창출해야 합니다.
   - **완전한 집합**: 리스크의 모든 차원을 포괄하며, 주어진 자산군의 체계적 리스크를 완전히 설명할 수 있는 집합을 형성해야 합니다.
2. **실제 상황에서의 팩터**:
   - **근사성**: 실제로는 이상적 조건이 근사적으로만 성립합니다.
       - '근사성'은 우리가 리스크 팩터를 이용해 시장을 분석하고 예측할 때, 완벽한 정확성이나 독립성을 기대할 수 없지만, 실제 시장 조건에 최대한 가까운 분석을 목표로 한다는 개념
   - **상관관계**: 다양한 팩터 간에 중요한 상관관계가 존재할 수 있습니다.
       - **예시**: 모멘텀은 소형주 사이에서 더 강하게 나타나는 경향이 있습니다.
3. **비지도학습을 이용한 합성 팩터**:
   - **주성분 분석(PCA)** 와 **독립 성분 분석(ICA)** 을 활용해 데이터 기반의 합성 팩터를 생성하는 방법을 배울 예정입니다.
4. **주요 팩터 범주 검토**:
   - 금융 연구와 트레이딩에서 유명한 몇 가지 주요 팩터 범주를 살펴보고, 경제적 논리와 수익률 동인을 설명합니다.
5. **팩터 구현과 평가**:
   - **넘파이**와 **판다스**를 사용한 팩터 구현 방법과 **TA-Lib**을 이용한 기술적 분석, **집라인 백테스트 라이브러리**를 통한 팩터 평가 방법을 알아봅니다.
   - **퀀토피안 플랫폼**에서 사용할 수 있는 내장된 팩터도 살펴볼 것입니다.

## 모멘텀과 심리 : 추세는 당신의 친구다

1. **모멘텀 투자 전략의 기본 원리**:
   - 이 전략은 과거에 성과가 좋았던 자산을 매수하고, 성과가 나쁜 자산을 매도하는 것에 기반합니다.
   - "추세는 너의 친구이다"라는 금언을 따릅니다.
2. **연구와 증거**:
   - Jagadeesh와 Titman의 연구부터 시작하여, 모멘텀 투자는 계량적 증거로 잘 뒷받침되고 있습니다.
   - AQR 헤지 펀드의 설립자 클리포드 애스니스는 다양한 자산 클래스와 시장에서 모멘텀 효과의 존재를 증명했습니다.
3. **전략의 전제**:
   - 자산 가격이 시간에 따라 긍정적인 방향으로 움직일 것이라는 추세를 반영합니다.
   - 이는 과거의 가격이 미래의 성능을 예측할 수 없다는 효율적 시장 가설과 대비됩니다.
4. **실제 적용**:
   - 가격 모멘텀은 여러 자산군에 걸쳐 긍정적인 수익률을 보여줍니다.
   - 많은 트레이딩 전략에서 중요한 역할을 합니다.
5. **모멘텀 팩터의 성과**:
   - 과거 2~12개월 동안 상위와 하위 성과를 기록한 미국 주식 사이의 성능 차이를 보여주는 '승자-패자' 포트폴리오를 통해 모멘텀 팩터의 효과를 입증합니다.
   - 2008년 금융 위기 이전까지 모멘텀 팩터는 다른 리스크 팩터들의 성과를 초과했습니다.

### 모멘텀과 심리가 초과 수익률을 낳는 이유

1. **행동 경제학적 이유**:
   - 투자자들이 시장 뉴스에 대해 과소반응하거나 과대반응하는 경향이 있습니다. 이는 시간이 지남에 따라 가격 모멘텀을 생성합니다.
   - 예를 들어, 기술주 랠리 같은 시장 버블은 과소반응과 이후의 과대반응의 결과입니다.
   - 공포와 욕망은 투자자들이 성능이 좋은 자산에 더 투자하고, 성능이 나쁜 자산을 팔도록 유도합니다.
2. **경제와의 양의 피드백 루프**:
   - 경제 성장은 주식시장을 상승시키며, 이는 다시 경제 성장을 촉진합니다.
   - 이러한 양의 피드백 루프는 주식과 신용 상품에서 모멘텀을 장기적으로 유지시킵니다.
3. **지속적인 수요와 공급의 불균형**:
   - 시장 마찰로 인해 생산이 수요 추세에 적응하는 데 시간이 걸리므로, 가격 상승 모멘텀이 발생할 수 있습니다.
       - 시장 마찰(Market Friction)이란 시장 참가자들이 자유롭게 거래할 수 있는 이상적인 시장 환경에서 벗어나, 실제 거래가 이루어지는 과정에서 발생하는 각종 비효율성이나 장애를 의미합니다
   - 예를 들어, 원유 생산이 수요를 따라잡는 데는 수년이 걸릴 수 있습니다.
4. **시장 미시 구조**:
   - 일정 기간 동안 투자자들이 편향을 모방하는 전략을 사용함으로써 가격 모멘텀을 생성합니다.
   - 손실을 줄이고 이익을 유지하는 전략, 포트폴리오 보험, 동적 델타 헤징, 방어적 풋과 같은 전략은 모멘텀을 강화합니다.
5. **리스크 패리티 전략**:
   - 변동성이 낮은 성과가 좋은 자산을 매수하고, 변동성이 높은 성과가 나쁜 자산을 매도하는 경향이 있습니다.
   - 이러한 전략들은 자동 리밸런싱을 통해 가격 모멘텀을 강화시킵니다.

### 모멘텀과 심리를 측정 하는 방법

1. **모멘텀 팩터의 기본**:
   - 가격 시계열 데이터에서 추세와 패턴을 식별하여 도출됩니다.
   - 이는 자산의 가격 변화를 기반으로 합니다.
2. **비교 방식**:
   - **전통 자산군 내 비교**: 같은 자산군 내에서 자산들을 서로 비교합니다.
   - **전통 자산군 간 비교**: 서로 다른 자산군 간의 비교를 통해 모멘텀을 식별합니다.
   - **횡단면적 비교**: 여러 기간에 걸친 자산들을 횡단면적으로 비교합니다. 이는 특정 시점에서 다양한 자산 간의 성능을 비교하는 것을 의미합니다.
   - **시계열 분석**: 하나의 자산의 시간에 따른 성능 변화를 분석합니다.
3. **수익률 기반 구축**:
   - **절대적 수익률**: 자산 자체의 성능을 기준으로 모멘텀을 측정합니다.
   - **상대적 수익률**: 다른 자산이나 벤치마크와 비교하여 자산의 성능을 측정합니다.
   
- 이 방법들은 모멘텀 투자 전략에서 중요한 역할을 하며, 투자자들이 가격 움직임의 추세를 파악하고 미래의 시장 방향성을 예측하는 데 도움을 줍니다. 모멘텀과 심리를 측정하는 이러한 접근 방식은 투자 결정 과정에서 중요한 정보를 제공하며, 다양한 시장 조건과 자산 클래스에서 효과적으로 적용될 수 있습니다.

## 가치 팩터

- 가치 팩터의 정의:
    - 기본적 가치에 비해 상대적으로 저평가된 주식이 시장 평균을 초과하는 수익률을 가져올 가능성이 높다는 원리에 기반합니다.
- 가치 전략의 원리:
    - 저렴하게 평가된 자산을 매수하고, 과대평가된 자산을 매도하는 신호를 제공합니다.
    - 자산의 공정 가치나 기본 가치를 추정하는 모델을 중심으로 합니다.

### 상대적 가치 전략

- 가격이 평균으로 회귀하리라 가정, 가격이 과대반응 또는 일시적 시장 충격으로 인해 공정 가치에서 벗어나는 경우에 초점을 맞춥니다.
- 가치주는 평균 회귀의 원리에 의존하며, 모멘텀 팩터와 반대의 특성을 가집니다.
- 가치 팩터의 구현:
    - 기본적 가치 전략과 시장 가치 전략을 포함한 다양한 전략으로 구현됩니다.
    - 통계적 차익 거래와 같이 시장 중립적인 롱/숏 포트폴리오로도 구현됩니다.

### 기본적 가치 전략

- 관련 경제적 및 기본적 지표를 사용하여 공정 자산 가치를 도출합니다.
- 채권, 통화, 상품 등 다양한 자산군에 적용됩니다.

### 시장 가치 전략

- 통계적 모델이나 머신러닝을 사용하여 유동성 공급의 비효율성으로 인한 가격 왜곡을 찾습니다.
- 단기적 시장 충격의 반전을 찾는 트레이딩 전략을 포함합니다.

### 자산 간 상대 가치 전략

- 자산 클래스 간의 상대적 가격 왜곡에 초점을 맞춥니다.
- 예를 들어, 전환 사채 차익 거래는 주식, 신용과 변동성 간의 상대적 가치를 거래합니다.

### 가치 팩터가 수익률을 예측하는 데 도움이 되는 이유
- 합리성 기반의 설명: 가치 프리미엄은 높은 실제 또는 인지된 리스크를 보상합니다.
- 행동 경제학적 설명: 손실 회피와 멘탈 회계 편향으로 인한 손실 회피와 할인율 조정에 의해 설명됩니다.
    - 멘탈 회계 편향(Mental Accounting Bias)은 개인이 경제적 결정을 내릴 때 자금을 가치가 다른 여러 '계정'으로 나누어 생각하는 심리적 경향
    - 투자자들은 이미 손실을 보고 있는 주식을 팔지 않고 보유하며, 반대로 이익이 난 주식을 너무 일찍 팔아 이익을 '잠그려고' 할 수 있습니다.
### 가치 효과를 포착하는 방법
- 기본 데이터에서 다양한 가치 대리 지표가 계산됩니다. 이 지표들은 자산의 가치를 평가하기 위한 기준으로 사용됩니다.

## 변동성과 규모 이례 현상

- 변동성과 규모가 투자 수익률에 미치는 영향과 이를 예측하는 다양한 방법을 탐색합니다. 저변동성 이례 현상과 규모 효과는 전통적인 금융 이론과 다르게, 낮은 리스크를 가진 자산이 때때로 높은 리스크를 가진 자산보다 더 좋은 성과를 낼 수 있다는 것을 보여줍니다. 이는 투자 전략과 자산 관리에 중요한 영향을 미칠 수 있습니다.

1. **규모 효과**: 낮은 시가 총액 주식이 높은 수익률을 제공하는 경향이 있습니다.
2. **저변동성 팩터**: 평균 이하의 변동성을 가진 주식이 더 높은 수익률을 제공합니다. 대형주는 일반적으로 변동성이 낮고, 전통적인 규모 팩터와 결합됩니다.
3. **역설**: 자본 자산 가격 모델(CAPM)은 높은 리스크가 높은 수익률을 가져와야 한다고 주장하지만, 실제로는 리스크가 낮은 자산이 때때로 더 좋은 성과를 보입니다.

### 변동성과 규모가 수익률을 예측하는 이유

1. **복권 효과**: 투자자들이 작은 손실 위험을 감수하면서도 큰 이익을 얻을 수 있는 주식에 끌립니다.
2. **대표 편향**: 변동성이 높은 주식의 성공 사례를 다른 모든 변동성이 높은 주식에 적용하는 경향이 있습니다.
3. **과신**: 투자자들은 자신의 예측 능력을 과신하며, 이는 변동성이 높은 주식의 가격을 부당하게 높일 수 있습니다.
4. **시장 기분**: 호황기에는 베타 차이가 크지 않지만, 위기 시에는 저변동성 주식이 선호되어 고변동성 자산보다 장기적으로 더 좋은 성과를 보입니다.

### 주요 척도

1. **저변동성 식별 척도**: 실현 변동성(표준 편차)과 예측(내재) 변동성을 포함합니다. 저변동성은 때때로 저베타로도 운용됩니다.
    - 주식이 과거나 미래에 얼마나 큰 움직임을 보일지를 측정해, 어떤 주식이 비교적 안정적인지를 파악하는 데 도움을 줍니다.
3. **증거의 강건성**: 저변동성 이례 현상에 대한 지지 증거는 여러 척도에 걸쳐 일관되게 나타납니다.
    - 여러 다른 방법으로 이 현상을 측정해 봐도 일관되게 '변동성이 낮은 주식이 장기적으로 더 좋은 성과를 낸다

## 우량주 팩터

- **목적**: 우량주 팩터는 고품질 기업, 즉 높은 수익성, 운영 효율성, 안전성, 좋은 지배 구조를 가진 기업에서 나오는 초과 수익을 찾는 것입니다.
- **시장 반응**: 시장은 일반적으로 확실한 이익을 가진 주식을 선호하고, 이익 변동성이 큰 주식은 저평가합니다.
- **투자자 선호**: 기본적 분석을 중시하는 투자자들에게 인기가 많지만, 계량투자자들에게는 상대적으로 새로운 개념입니다.
- **전략적 활용**: 우량주 팩터는 다른 리스크 팩터, 특히 가치 팩터와 결합하여 멀티팩터 전략의 질을 높이는 데 자주 사용됩니다.

### 우량주 팩터가 중요한 이유

- **장기적 성과**: 우량주는 지속 가능한 이익, 안정적인 현금 흐름, 적절한 레버리지, 낮은 재무 리스크를 가지며, 이는 장기적으로 주가를 지지합니다.
- **행동 경제학적 요인**: 투자자들은 종종 기업의 질에 대한 정보에 과소 반응하며, 우량주에 대한 선호로 이어집니다.
    - 초기에는 정보에 대한 과소 반응이 있지만, 이러한 과소 반응이 장기적으로 우량한 기업의 주식에 대한 관심과 수요를 증가시키며, 결국 이러한 기업의 주식이 시장 평균을 상회하는 성과를 낼 수 있다는 것
- **투자자 정당화**: 펀드 매니저들은 강한 기본을 가진 비싼 주식을 구매하는 것이 정당화하기 더 쉬워, 우량주 프리미엄이 발생합니다.

### 주요 척도

- **우량성 측정**: 대차대조표와 손익계산서에서 파악된 높은 수익이나 현금 흐름 마진, 운영 효율성, 재무 건전성 등을 통해 우량성을 측정합니다.
- **척도**: 매출 총이익, 투자 자본 수익률, 이익 변동성 등 다양한 척도를 사용하여 우량주를 식별합니다.
- **이익의 질**: 발생액의 크기를 통해 이익의 질을 평가하며, 큰 발생액은 이익의 질이 낮을 가능성을 시사합니다.

### 여러 과거 기간을 위한 수익률 계산

- **역사적 수익률 계산**: `.pct_change(n_periods)` 메소드를 사용하여, 다양한 월간 기간 동안의 수익률을 계산합니다. 이 과정은 모멘텀과 같은 시계열 패턴을 반영합니다.
- **데이터 형식 변환**: 계산된 수익률 데이터를 `.stack()` 메소드를 통해 긴 형식으로 전환합니다.
- **수익률의 이상치 처리**: `.pipe()`와 `.clip()`을 사용하여, 수익률 데이터의 이상치를 [1%, 99%] 수준에서 제한하여 윈저화합니다.
- **수익률 정규화**: 기하평균을 사용하여 수익률을 정규화합니다.
- **복리 월간 수익률 계산**: `.swaplevel()`을 사용하여 MultiIndex 레벨의 순서를 바꾼 후, 1개월에서 12개월까지 여러 기간에 대한 복리 월간 수익률을 계산합니다.

```python
outlier_cutoff = 0.01
data = pd.DataFrame()
lags = [1, 2, 3, 6, 9, 12]
for lag in lags:
    data[f'return_{lag}m'] = (monthly_prices
                           .pct_change(lag)
                           .stack()
                           .pipe(lambda x: x.clip(lower=x.quantile(outlier_cutoff),
                                                  upper=x.quantile(1-outlier_cutoff)))
                           .add(1)
                           .pow(1/lag)
                           .sub(1)
                           )
data = data.swaplevel().dropna()
data.info()
```

## 칼만 필터를 사용한 알파 팩터의 잡음 제거

- **잡음의 개념**: 데이터에서 '잡음'은 원하지 않는 정보나 신호의 방해를 의미합니다. 신호 처리 분야에서는 원래의 순수한 신호에서 이러한 잡음을 분리하는 것이 중요한 과제입니다.
- **칼만 필터의 역할**:
   - 1960년에 도입된 칼만 필터는 잡음이 많은 데이터에서 원래 신호의 더 정확한 추정치를 제공하는 데 사용됩니다.
   - 이 기법은 시계열 분석, 컴퓨터 비전, 항공항로 추적, 로봇 제어 등 다양한 분야에서 널리 적용됩니다.
- **금융에서의 응용**:
   - 금융 데이터 분석에서도 칼만 필터는 유사한 방식으로 사용됩니다. 특히, 트레이딩 신호를 생성할 때 원하지 않는 잡음을 제거하고, 유용한 정보만을 추출하는 데 도움을 줍니다.
   - 잡음과 유용한 신호를 분리하는 것은 어려운 과제일 수 있지만, 칼만 필터를 사용하면 이러한 과정이 효율적으로 이루어질 수 있습니다.
- 칼만 필터는 데이터에서 원하지 않는 잡음을 제거하고 순수한 신호를 추출하는 데 매우 유용한 도구입니다. 금융 데이터 분석 및 트레이딩 신호 생성에서 칼만 필터를 사용함으로써 보다 정확한 정보에 기반한 결정을 내릴 수 있으며, 이는 알파 팩터의 성능을 개선하는 데 기여할 수 있습니다.

### 칼만 필터의 작동 원리

<img src="./images/fig_04_05.png" width=700>

1. **예측 단계**: 프로세스의 현재 상태를 기반으로 미래 상태를 추정합니다.
    - 예측하기: 칼만 필터는 현재 상태를 바탕으로 무슨 일이 일어날지 추측합니다. 예를 들어, 당신이 어디로 갈지 예측하는 거죠.
2. **측정 단계**: 실제 관측치를 사용하여 예측을 업데이트하고, 더 확실한 추정치에 더 많은 가중치를 줍니다.
    - 측정하기: 실제로 일어난 일을 관찰합니다. 예를 들면, 실제로 어디에 있는지 확인하는 것이죠.
    - 업데이트하기: 처음에 예측했던 것과 실제로 일어난 일을 비교해서, 다음 예측을 더 정확하게 만듭니다.
#### 칼만 필터의 가정
- 시스템이 선형으로 작동합니다.
    - 선형성: 칼만 필터는 모든 것이 직선(선형)으로 예측된다고 가정합니다. 즉, 현재 상태에서 미래 상태로 가는 변화가 직선적인 패턴을 따른다는 것이죠.
- 은닉 상태 프로세스는 현재 상태가 바로 이전 상태에만 의존하는 마르코프 체인입니다.
    - 마르코프 체인: 이는 현재의 상태가 오직 바로 이전 상태에만 의존한다는 뜻입니다. 과거의 모든 것을 잊고, '바로 전에 무슨 일이 있었나?'만 기억하죠.
- 측정치는 고정된 공분산을 가진 가우시안(정규) 잡음에 의해 영향을 받습니다.
    - 가우시안 잡음: 관측치(우리가 볼 수 있는 데이터)는 예측하기 어려운 '잡음'에 의해 영향을 받습니다. 이 '잡음'은 정규 분포(가우시안)를 따른다고 가정합니다.
#### 모델 구성 요소
- **초기 상태**: 정규 분포 $N(\mu_0, \sigma_0^2)$를 따릅니다.
    - 시작할 때, 우리는 시스템이 어떤 상태에 있는지 최선의 추측을 합니다. 이 추측은 정규 분포를 따릅니다.
- **전이(상태 변환)**: 다음 상태는 전이 행렬, 전이 편향, 추가된 가우시안 잡음을 포함한 선형 변환으로 표현됩니다.
    - 시간이 지남에 따라, 시스템의 상태가 어떻게 변하는지를 설명합니다. 이 변화는 전이 행렬(방향을 결정하는 규칙), 전이 편향(조금의 조정), 그리고 '잡음'을 포함합니다.
- **관측치**: 관측 행렬, 관측 편향, 추가된 가우시안 잡음을 포함한 은닉 상태의 선형 변환입니다.
    - 우리가 실제로 볼 수 있는 것, 즉 시스템의 현재 상태에 대한 관찰입니다. 이 관찰 역시 '잡음'에 의해 영향을 받습니다.
#### 장단점
- **이점**: 비정상성 데이터에 유연하게 적응합니다.
- **단점**: 선형성과 가우시안 잡음의 가정을 기반으로 하며, 이는 금융 데이터에서 자주 위배됩니다. 비선형 시스템에는 확장 칼만 필터나 무향 칼만 필터, 비정규 분포 추정에는 입자 필터를 사용할 수 있습니다.

## 웨리블릿을 이용한 잡음 데이터의 전처리 방법

- **웨이블릿이란?**
    - 웨이블릿은 여러 주파수의 사인과 코사인 파동을 결합하여 잡음이 섞인 신호를 근사화하는 방법입니다. 푸리에 분석과 관련이 있지만, 특정 시간대의 신호 변화를 더 잘 포착합니다.
    - 웨이블릿은 마치 신호 속에서 숨은 그림을 찾아내는 마법 같은 도구예요. 신호란 음악이나 주식 가격 같이 시간에 따라 변하는 어떤 정보를 말합니다. 잡음이란 우리가 원하는 정보를 방해하는 불필요한 데이터입니다. 웨이블릿은 이 잡음을 걸러내고 우리가 찾고 싶은 정보만 남겨주는 역할을 합니다.
- **웨이블릿의 작동 방식**:
    - 웨이블릿 변환은 신호를 여러 크기와 위치의 파동 형태로 분해하여, 각각의 파동이 신호의 다른 부분을 나타내게 합니다. 이는 푸리에 변환보다 불규칙하고 비주기적인 신호를 더 잘 처리할 수 있습니다.
    - 웨이블릿은 신호를 작은 파동들로 나누어 분석합니다. 마치 음악을 다양한 악기 소리로 나누어 듣는 것과 비슷해요.
    - 그리고 각 파동이 신호의 어떤 부분을 나타내는지를 보여줍니다. 이 과정을 통해, 우리는 신호 속에서 중요한 부분을 쉽게 찾아낼 수 있습니다.
- **잡음 제거 과정**:
    1. **신호 분해**: 웨이블릿은 신호를 여러 조각으로 나눕니다. 이 조각들은 신호의 다른 특성을 나타냅니다.
    2. **중요한 부분 선택**: 중요하다고 생각되는 조각들만 골라냅니다. 나머지는 잡음일 가능성이 높기 때문에 무시합니다.
    3. **신호 복구**: 중요한 조각들만 가지고 원래의 신호를 다시 만듭니다. 이때 잡음은 대부분 제거됩니다.

- 웨이블릿은 음악이나 영상에서 잡음을 제거하는 데 사용될 수 있으며, 금융 데이터에서 중요한 경향을 찾아내거나, 심지어 의료 영상에서 중요한 정보를 찾아내는 데도 사용됩니다.

## 다양한 데이터 소스의 팩터 결합

```python
from zipline.api import order, record, symbol
from zipline.algorithm import TradingAlgorithm
import pandas as pd
import numpy as np
from datetime import datetime

# 이동평균 크로스오버 전략 정의
def initialize(context):
    context.asset = symbol('AAPL')
    context.short_window = 50
    context.long_window = 200

# 매수 및 매도 로직 정의
def handle_data(context, data):
    short_mavg = data.history(context.asset, 'price', bar_count=context.short_window, frequency="1d").mean()
    long_mavg = data.history(context.asset, 'price', bar_count=context.long_window, frequency="1d").mean()

    if short_mavg > long_mavg:
        order(context.asset, 100)
    elif short_mavg < long_mavg:
        order(context.asset, -100)

# 백테스트 시작과 종료 날짜 설정
start_date = pd.Timestamp(datetime(2010, 1, 1))
end_date = pd.Timestamp(datetime(2020, 12, 31))

# 주식 데이터 불러오기 (여기서는 AAPL 주식 데이터 사용)
data = pd.read_csv("AAPL.csv", index_col="date", parse_dates=True)

# Zipline 백테스트 알고리즘 실행
algo = TradingAlgorithm(initialize=initialize, handle_data=handle_data)
results = algo.run(data.loc[start_date:end_date])

# 백테스트 결과 출력
print(results)

# 결과 그래프 생성
import matplotlib.pyplot as plt
results.portfolio_value.plot()
plt.show()
```

  - 벤치마크가 무위험 이자율일 경우, IR은 샤프 비율과 유사합니다.
- **적극적 운용의 근본적 법칙**:
  - 이는 IR을 예측 기술과 해당 예측을 활용하는 전략의 능력으로 분해합니다. 예측 기술이란 미래 수익률의 방향성을 얼마나 잘 예측하는지를 말합니다.
- **성과 척도**:
  - 알파 팩터의 예측과 실제 미래 수익률 사이의 상관관계를 기반으로 합니다.
  - **스피어맨 순위 상관계수**를 사용하는 것이 바람직한데, 이는 두 변수 간의 관계를 단조 함수로 측정하여, 피어슨 상관계수가 선형 관계만 측정하는 것과 대비됩니다.

# 알파 팩터 리소스
#### 신호에서 트레이딩으로: 알파 팩터 설계 및 선정
- **핵심 목표**: 머신러닝(ML) 모델을 이용해 신호의 예측력에 중점을 둔 알파 팩터를 설계하고 선정합니다.
- **전략 기반**: 여러 개의 알파 팩터를 사용해 각 종목에 대한 신호를 제공하고, 이를 통해 포지션의 시기와 크기에 대한 결정을 최적화합니다.

## 기타 알고리즘 트레이딩 라이브러리
1. **QuantConnect**
   - Quantopian의 경쟁 플랫폼으로, 다양한 알고리즘 트레이딩 전략 개발과 백테스트를 지원합니다.
2. **WorldQuant**
   - 온라인 경연대회를 제공하며, 크라우드 소싱을 통한 헤지 펀드에 커뮤니티 공헌자들을 채용합니다.
3. **Alpana Trading Labs**
   - Quantopian과 유사한 비즈니스 모델을 가지며, 고빈도 트레이딩에 초점을 맞춘 테스트 인프라를 제공합니다.
4. **Python Algorithmic Trading Library (PyAlgoTrade)**
   - 백테스트에 중점을 둔 라이브러리로, 가상 트레이드와 실전 트레이드를 지원하여 사용자가 트레이딩 전략을 쉽게 평가할 수 있게 합니다.
5. **Pybacktest**
   - 벡터화된 백테스트 프레임워크이며, 판다스를 사용하여 간결하고 빠른 백테스트를 지향합니다. (현재 프로젝트 중단)
6. **UltraFinance**
   - 실시간 재무 데이터 수집과 트레이딩 전략의 분석 및 백테스트를 결합한 오래된 프로젝트입니다.
7. **Trading with Python**
   - 퀀트 트레이딩을 위한 교과 과정, 함수 모음집 및 수업을 제공합니다.
8. **Interactive Brokers**
   - 실시간 트레이딩을 위한 파이썬 API를 자신들의 플랫폼에서 제공합니다.
9. **Backtrader**
   - 파이썬 기반의 백테스트 및 알고리즘 트레이딩 프레임워크로, 다양한 데이터 피드와 전략을 지원합니다.

# 요약
- **알파 팩터 소개**: 수십 년 간의 전략 설계 및 평가 방법을 통해 알려진 경제적 메커니즘을 바탕으로 알파 팩터의 작동 원리와 초과 수익률 산출 원리에 대해 설명합니다.
- **도구 제시**: 다양한 데이터 소스를 이용하여 개인의 알파 팩터를 생성할 수 있는 도구를 소개하고, 잡음이 있는 데이터를 처리하는 칼만 필터와 웨이블릿 방법을 시연합니다.
- **집라인 라이브러리**: 알고리듬 트레이딩의 이벤트 기반 시뮬레이션을 위해 사용되며, 알파렌즈 라이브러리를 통해 신호의 예측 성과 및 거래 회전율을 평가할 수 있습니다.
- **백테스트 문제**: 과거 데이터 세트에 대한 트레이딩 전략의 백테스트 문제와 해결 방법에 대해 논의합니다.
