## 투자 시 세금과 비용에 대한 생각

- 세금과 비용은 중요
  - 세금과 비용은 장기 수익율에 큰 영향을 미친다.
  - 장기 투자 시 세금을 이연할 수 있는 절세 계좌 효용이 좋다.
  - 하지만, 수익을 내는 것이 더 중요
- 거래 비용은 가능한 한 낮추자
  - 잦은 거래는 거래 비용을 크게 높인다.
  - 증권사 수수료는 비교해서 가능한한 낮추자.

### 개인적인 생각
- 국내 주식보다는 미국 주식 위주 투자를 선호
  - 국내 주식이 변동성은 크고 이익률은 떨어진다. (국민연금)
    - https://signalm.sedaily.com/NewsView/2DABLNOPDV/GX1501
  - 장기 투자로 S&P500, 나스닥 100 지수 투자 선호
- 미국 주식 투자 시 투자 이익이 어느 이상 크기가 되면 양도소득이 배당소득보다 유리하다
  - 배당소득은 커지면 누진 과세로 과세율이 올라간다.
  - 배당소득은 주기적 소득으로 간주되어 준조세가 발생한다.

## 주식 투자 시 기본 세금 및 비용

국내 주식의 경우 금융투자소득세를 적용하지 않을 경우

### 국내 주식 투자 시

- 증권 거래세: 0.18 %
- 거래 수수료
  - 유관기관 수수료 0.0036396 %
  - 나머지 수수료는 평생 무료 혜택 이벤트를 활용
- 매매 차익: 0 %
  - 대주주 요건에 해당하면 22 % (한 종목 50억(현재 - 변경 가능성 높음) 이상)
- 배당 이익: 15.4 % (14 % + 1.4 %) (원천징수, 배당소득세)
 
### 국내 ETF 투자 시

- 증권 거래세가 부과되지 않는다.
- Max(매매 차익, 과표 기준 차익)의 15.4 % 과세 (원천징수)
  - ETF 내에서 발생하는 국내 주식 매매 차익은 과세되지 않는다는 뜻

### 미국 주식 직접 투자 시

- SEC fee (매도 만): 0.00278 %
- 거래 수수료
  - 증권사 별로 차이가 크다
  - 삼성증권: 0.03%
  - 한국투자증권: 0.05%
  - 토스: 0.1%
    - 대신 소수점 거래도 같은 수수료율 (타 증권사는 소수점 거래 수수료가 높다.)
    - $10 미만 매수와 '주식 모으기'로 매수는 수수료 0 %
- 매매 차익: 22 % (20% + 2%) 
- 배당 이익: 15 % (미국 내 원천징수, 배당소득세)
  - 일부 미국 외 주식은 현지 과세 규정에 따라 과세
    - Ex)
      - EQNR(에퀴노르 ADR): 노르웨이 25 %
      - ZIM: 이스라엘 25 %
    - 조세조약에 따른 제한 세율을 적용받을 수 있으나 비용과 노력이 든다.

### 해외 주식 국내 상장 ETF 투자 시

- Max(매매 차익, 과표 기준 차익)의 15.4 % 과세 (원천징수, 배당소득세)
  - 일부 비과세 이익을 빼고 주식 양도 차익도 배당 소득세로 과세된다고 생각하면 됨.
- 분배금: 15.4 % 과세 (원천징수, 배당소득세)

## 부가적으로 발생하는 세금, 준조세, 비용

### 금융소득 종합과세

- 2000 만원이 넘는 금융소득은 종합소득세 신고 시 합신 신고해야함
  - 배당 소득 + 이자 소득
- 종합소득세 신고 시 과세표준이 약 7900만원이 넘을 때 실질 과세 금액이 발생
  - 약 7900만원이 넘어야 총 과세율이 15.4 %를 넘는다.
    
### 건강보험 보수 외 소득월액 보험료

- 2000 만원이 넘는 기타 소득이 있을 경우 초과분에 대해사 건강보험료가 부과됨
  - 분리과세 임대 소득도 포함
  - 1000 만원 ~ 2000 만원의 분리과세 금융 소득도 포함
- 소득월액 보험료율: 7.09 % + 0.9182 %
  - 소득월액이 부과되는 금액에 대해서는 <span style="color:red">8.0082 %</span> 의 준조세 발생 (15.4 % + 8.0082 % = 23.4082 %)
  - 매년 보험료율이 올라감.
- https://www.nhis.or.kr/nhis/policy/wbhada01900m01.do

### 양도소득 

- 250만원까지는 공제됨.

## 배당소득을 줄이는 방법

해외 투자를 할 때는 배당소득과 양도소득이 서로 Trade-off 가능하다. 
- 임대소득이 없을 경우 2000만원, 임대소득이 있을 경우 1000만원이 넘을 때 부터 건강보험 보수 외 소득월액이 부과될 수 있다.
- 이자-배당소득이 늘어서 부담될 경우 양도소득에 더 집중할 수 있다.

### Total Return ETF에 투자

배당이 나올 경우 분배하지 않고 자동으로 재투자하는 경우 Total Return이라고 한다. 배당을 하지 않는 대신 ETF 가격이 올라감.

- 한국에 상장된 TR ETF에 투자할 수 있다.
  - 이름이 TR로 끝나는 ETF들
  - 금융투자소득세가 적용되면 모두 없어질 수 있다.
  - 매도 시 이익의 대부분이 배당소득세로 과세된다. (대신 시점 조절 가능)
  - Ex)
    - Kodex 미국 S&P500TR
    - Kodex 미국 나스닥100TR
- 미국에 상장된 ETF는 TR이 거의 없다.
  - 세금 문제로 TR로 ETF를 만들기기 어렵다.
- 유럽에 상장된 미국 주식 ETF를 투자할 수 있다.
  - 미국의 유명 ETF들이 유럽 시장에 동반 상장되어 있음.
  - 영국 시장의 경우 제목에
    - UCITS이 포함되면 미국 ETF
    - Acc가 포함되면 Total Return
    - USD가 포함되면 달러로 매매
  - Ex) 영국 시장에 상장된 것들
    - iShares NASDAQ 100 UCITS ETF USD (Acc) (CNDX)
    - Vanguard S&P 500 UCITS ETF USD Acc (VUAA)
    - iShares $ Treasury Bond 0-1yr UCITS ETF USD A (IB01)
 
### 배당이 적은 기업에 투자

배당을 하지 않고 재투자나 자사주 매입 및 소각을 하는 기업에 투자할 수 있다.

Ex)
- 버크셔 해서웨이 Class B
  - 장기적 우상향하면서 배당을 하지 않는다.
  - 보험, 철도, 에너지, 등등..

## 법인 설립

앞의 방법 외에 절세가 가능한 방법으로는 법인 설립이 있다.

방법
- 1인 법인을 설립한다.
- 법인용으로 증권사 계좌를 만들고 해당 계좌에서 투자

장점
- 낮은 세금
  - 2억까지 9% 세율
- 손익 통산
  - 같은 해에 발생한 투자 손실과 이익을 서로 상계할 수 있다.
  - 5년 내에 발생한 이월결손금을 공제할 수 있다.

단점
- 높은 수수료
  - 법인 계좌를 다루는 증권사가 제한되고 수수료율도 높다.
    - 한국투자증권의 경우 
      - 국내 0.015%
      - 해외 0.07%
- 회계 처리
  - 모든 거래 내역을 회계 장부를 작성해서 세금 처리를 해서 법인세 신고 필요
  - 법인세 성실신고확인 대상이 될 수 있다.
    - 세무서에 맡길 경우 수수료 발생 (약 100만원?)
    - 성실신고 하지 않을 경우 법인세 산출세액*의 5%와 수입금액의 0.02% 중 큰 금액이 가산세
- 법인 업무 처리
  - 급여 등 처리
  - 등기 관련 처리
  - ...
- 법인 돈을 개인이 받으면 소득세 발생
  - 법인 돈은 개인이 그냥 사용할 수 없음
  - 배당으로 받아서 배당소득세를 내는 방법
    - 15.4 % 배당소득세 납부
    - 총 23.014 % 세금
  - 급여로 받아서 근로소득세를 내는 방법
    - 기존 급여가 있을 경우 종합소득세 신고 때 합산
    - 겸업금지 문제가 발생 가능
    - 기존 직장이 없을 경우 낮은 근로소득세가 가능하다.
