# Intro

### 선형 모델의 중요성
- **적용의 넓이**: 선형 모델은 금융 데이터에 대한 분석에서 널리 사용되며, 그 이유는 훈련이 효율적이고, 금융 이론과의 연결이 강하며, 데이터의 잡음에 강한 구조를 가지고 있기 때문입니다.
- **직관적이고 해석 가능**: 선형 예측 변수는 해석하기 쉽고 데이터를 합리적으로 잘 적합시키는 데 도움을 줍니다.

### 선형 회귀의 발전
- **역사적 배경**: 선형 회귀는 200년 이상의 역사를 가지고 있으며, 최소 자승법을 사용하여 파라미터를 학습하는 방법으로 발전해 왔습니다.
- **일반화 선형 모델 (GLM)**: 정규 분포가 아닌 다른 오차 분포를 가진 데이터에도 적용할 수 있도록 확장된 모델입니다. 범주형 반응 변수를 위한 로지스틱 회귀 등이 포함됩니다.

### 선형 모델의 확장 기법
- **강건한 추정 방법**: 데이터가 일반 가정을 위반하는 경우에도 통계적 추론을 가능하게 합니다. 예를 들어, 패널 데이터에서 반복적으로 관찰되는 경우에 유용합니다.
- **수축 방법**: 모델의 예측 성능을 향상시키기 위해 사용되며, 모델 계수에 복잡성 페널티를 부여하여 모델의 분산을 줄이고 예측 성능을 개선합니다.

### 선형 모델의 응용
- **회귀와 분류**: 다양한 시간에 걸친 수익률 예측, 리스크 관리, 방향성 가격 예측 등 다양한 금융 문제에 활용됩니다.
- **산업 및 학계 연구**: 자산 가격 결정 모델에서 널리 사용되며, 학계와 산업 연구자들이 활발히 개발하고 응용하고 있습니다.

# 추론에서 예측까지

#### 기본 원리
- **선형 결합 가정**: 선형 회귀는 변수들의 선형 결합을 통해 예측을 수행하며, 관측치의 오차를 확률적으로 허용합니다.
- **오차의 독립성과 동질성**: 오차는 독립적이고 일정한 분산을 가진다고 가정합니다. 이는 계산을 단순화하고, 오차가 서로 영향을 미치지 않음을 의미합니다.

#### 최소 자승법 (OLS)
- **효율적인 추정 제공**: 최소 자승법은 편향이 없고 가장 낮은 분산을 가지는 효율적인 추정치를 제공합니다.
- **실제 데이터의 문제점**: 많은 경우, 실제 데이터는 이상적인 가정을 만족시키지 못하고 오차의 독립성이 위배될 수 있습니다.

#### 패널 데이터와 복잡한 구조
- **패널 데이터의 특성**: 시간에 걸쳐 같은 단위의 반복 관측이 포함된 데이터로, 상관관계가 포함될 수 있습니다.
- **오차 구조의 복잡성**: 실제 금융 데이터는 종종 복잡한 오차 구조를 가지며, 이를 모델링하기 위해 더 정교한 통계적 기법이 필요합니다.

#### 수축 방법과 규제화
- **모델의 복잡성 감소**: 수축 방법은 모델 계수에 페널티를 부여하여 과적합을 방지하고, 예측 성능을 향상시킵니다.
- **규제화의 역할**: 규제는 편향과 분산의 최적 균형을 찾아 예측 성능을 극대화합니다.

#### 선형 모델의 활용
- **자산 수익률 예측**: 다양한 시간 구간에 대한 자산 수익률을 예측하는 데 선형 모델이 널리 사용됩니다.
- **알고리즘 트레이딩 전략**: 선형 모델은 알고리즘 트레이딩에서 중요한 팩터 모델 추정과 거래 신호 생성에 활용됩니다.

# 베이스라인 모델 : 다중 선형 회귀 모델

## 모델 만드는 방법

#### 기본 개념
- **다중 회귀 모델**: 하나의 연속적인 결과 변수와 여러 입력 변수 간의 선형 관계를 모델링합니다. 입력 변수는 전처리가 필요할 수 있습니다.
- **다변량 회귀 모델**: 여러 입력 변수에 대한 여러 출력 변수의 관계를 분석하는 회귀 모델입니다.

#### 모델 구조
- **수식**: 모집단에서 선형 회귀 모델은 일반적으로 다음과 같은 형식을 가집니다:
  $$
  Y = f(X) + E = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_p x_p + E
  $$
  여기서 $Y$는 출력, $X$는 입력 벡터, $E$는 오차, $\beta_0, \beta_1, \dots, \beta_p$는 모델 계수입니다.

- **행렬 표현**:
  $
  \mathbf{Y} = \mathbf{X}\beta + \mathbf{E}
  $
  - $ \mathbf{Y} $는 $N$개의 출력 관측치를 포함하는 벡터 (Nx1)
  - $ \mathbf{X} $는 관측치와 상수항의 열을 포함하는 계획 행렬 (NXP)
  - $ \beta $는 계수 벡터 (PX1)
  - $ \mathbf{E} $는 오차 벡터 (NX1)

#### 계수의 해석
- 각 계수 $\beta_j$는 다른 모든 변수가 고정된 상태에서 해당 변수 $x_j$가 출력 $Y$에 미치는 부분적인 평균 효과를 나타냅니다.

#### 모델의 확장
- **비선형 관계**: 변수 변환(예: 로그 변환, 다항식)을 통해 모델에 비선형성을 도입할 수 있습니다.
- **더미 인코딩**: 범주형 변수를 포함하기 위해 더미 변수를 사용할 수 있습니다.
- **상호작용 효과**: 두 변수의 상호작용을 나타내기 위해 새로운 변수를 생성할 수 있습니다.

#### 통계적 접근
- **가설 검증**: 모델을 구성할 때는 계수에 대한 가설을 검증하고, 오차 항에 관한 특정 가정(독립성, 동질성)을 설정하는 것이 중요합니다.
- **파라미터 학습**: 데이터로부터 파라미터를 학습시키기 위한 통계적 방법을 적용하고 이를 수행합니다.

## 모델을 훈련하는 방법

### 최소 자승법 : 초평면을 데이터에 맞추는 방법

- 최소 자승법(Ordinary Least Squares, OLS)은 선형 회귀 모델을 훈련시키는 표준적인 방법입니다.
- 이 방법은 데이터의 예측값과 실제값 사이의 차이, 즉 잔차의 제곱합을 최소화하여 모델 파라미터를 추정합니다.

#### 최소 자승법의 기본 원리
- 최소 자승법은 다음과 같은 선형 관계를 모델링하는 데 사용됩니다:
- $ Y = X\beta + \epsilon $
    - $ Y $는 응답 변수,
    - $ X $는 설명 변수(독립 변수),
    - $ \beta $는 추정해야 할 모델 파라미터,
    - $ \epsilon $은 오차 항입니다.

#### 목표
- 최소 자승법의 주된 목표는 잔차 제곱합(Residual Sum of Squares, RSS)을 최소화하는 것입니다:
- $ \text{RSS}(\beta) = (Y - X\beta)^\top (Y - X\beta) $

#### 최적화 문제 풀이
- RSS를 최소화하기 위해 $ \beta $에 대해 미분한 후 결과를 0으로 설정합니다:
- $ \frac{\partial}{\partial \beta} \text{RSS}(\beta) = -2X^\top(Y - X\beta) = 0 $
- 위 식을 재배열하여 $ \beta $를 구합니다:
- $ X^\top X\beta = X^\top Y $
- $ \beta = (X^\top X)^{-1}X^\top Y $

#### 기하학적 해석
- 최소 자승 추정치는 $ X $의 열공간에 대한 $ Y $의 직교 투영입니다. 잔차 벡터 $ Y - \hat{Y} $는 $ X $의 열공간에 직교합니다.

#### 가정
- 최소 자승법은 입력 변수가 선형 독립이고, 오차가 정규 분포를 따르며, 독립적이고 등분산성을 가진다는 가정 하에 사용됩니다. 이러한 가정이 충족되지 않을 때 다른 통계적 방법이나 모델이 고려될 수 있습니다.

### 최대 우도 추정

- 최대 우도 추정(Maximum Likelihood Estimation, MLE)은 주어진 데이터에 대한 모델의 파라미터를 추정하는 통계적 방법입니다. 이 방법은 주어진 파라미터가 데이터를 생성할 가능성(우도)을 최대화하는 값을 찾습니다.

#### 최대 우도 추정의 기본 개념
- 최대 우도 추정은 모델 파라미터를 사용하여 관찰된 데이터가 나타날 우도를 계산하는 우도 함수에 기반을 둡니다. 우도는 확률과 달리 0과 1 사이로 정규화되지 않으며, 각 데이터 포인트의 확률을 곱하여 전체 데이터 세트의 우도를 구합니다.

#### 우도 함수
- 선형 회귀의 경우, 오차가 정규 분포를 따른다고 가정할 때, 우도 함수는 다음과 같이 정의됩니다:
- $ P(Y_i | X_i, \beta) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(Y_i - X_i^T \beta)^2}{2\sigma^2}\right) $
    - $ Y_i $는 응답 변수,
    - $ X_i $는 설명 변수,
    - $ \beta $는 모델 파라미터,
    - $ \sigma^2 $는 오차의 분산입니다.

#### 로그 우도 함수
- 곱셈 형태의 우도 함수는 계산상의 어려움을 야기하기 때문에, 로그 우도 함수를 사용하여 더 쉽게 최적화합니다:
- $ \log L(\beta) = -\frac{n}{2} \log(2\pi\sigma^2) - \frac{1}{2\sigma^2} \sum_{i=1}^n (Y_i - X_i^T \beta)^2 $
- 여기서 $ n $은 샘플의 개수입니다.

#### 파라미터 추정
- 최대 우도 추정의 목표는 로그 우도 함수를 최대화하는 파라미터 $ \beta $를 찾는 것입니다. 이를 통해 파라미터 추정치는 다음과 같이 계산됩니다:
- $ \hat{\beta}_{MLE} = \arg \max_{\beta} \log L(\beta) $
- 정규 분포의 오차를 가정하면, 이 최적화 문제는 최소제곱법과 동일한 결과를 산출합니다.

#### 일반화
- MLE는 다양한 오차 분포와 복잡한 모델에 적용될 수 있습니다. 예를 들어, 로지스틱 회귀에서는 베르누이 분포를 사용하여 이진 분류 문제를 모델링하며, 최소제곱법을 사용할 수 없는 경우에 MLE가 효과적인 대안이 됩니다.

- 최대 우도 추정은 그 유연성과 강력한 통계적 기반으로 인해 다양한 모델링 상황에서 널리 사용됩니다.

### 경사 하강법

- 경사 하강법(Gradient Descent)은 다양한 최적화 문제에서 널리 사용되는 방법으로, 특히 머신러닝과 딥러닝에서 모델의 파라미터를 학습할 때 기본적으로 적용됩니다. 이 방법은 함수의 최소값을 찾기 위해 반복적으로 파라미터를 조정해 나갑니다.

#### 경사 하강법의 기본 개념
- 경사 하강법은 함수의 기울기(그래디언트)를 계산하여, 그 기울기가 가리키는 방향(즉, 가장 빠르게 증가하는 방향의 반대 방향)으로 일정한 크기의 스텝을 반복적으로 이동함으로써 함수의 최소값을 찾습니다.

#### 경사 하강법의 수학적 표현
- 목적 함수 $ f(\theta) $에 대해, 파라미터 $ \theta $의 값을 업데이트하는 기본 수식은 다음과 같습니다:
- $ \theta := \theta - \alpha \nabla_\theta f(\theta) $
    - $ \alpha $는 학습률(learning rate)로, 각 스텝에서 파라미터를 얼마나 조정할지를 결정합니다.
    - $ \nabla_\theta f(\theta) $는 $ \theta $에서 함수의 그래디언트입니다.

#### 알고리즘의 변형
1. **배치 경사 하강법**: 전체 데이터셋을 사용하여 한 번에 하나의 스텝을 계산합니다.
2. **확률적 경사 하강법 (SGD)**: 데이터셋에서 무작위로 선택된 하나의 데이터 포인트를 사용하여 스텝을 계산합니다. 이 방법은 더 빠른 수렴을 가능하게 하지만 더 많은 변동성을 가집니다.
3. **미니배치 경사 하강법**: 데이터셋의 작은 부분집합(미니배치)을 사용하여 각 스텝을 계산합니다. 이 방법은 배치와 확률적 경사 하강법의 장점을 결합합니다.

#### 수렴 조건
1. **학습률 $ \alpha $**: 너무 크면 수렴하지 않고 발산할 수 있으며, 너무 작으면 수렴 속도가 매우 느려집니다.
1. **허용오차(tolerance)**: 이 값 이하로 손실이 개선되지 않을 때 알고리즘을 중단할 수 있습니다.
1. **에폭(epoch)**: 전체 데이터셋에 대한 단일 반복을 의미하며, 수렴할 때까지 여러 에폭을 실행할 수 있습니다.

- 경사 하강법은 그 구현의 단순성과 범용성으로 인해 매우 인기 있는 최적화 도구입니다. 그러나 적절한 학습률 설정과 수렴 조건을 정의하는 것이 중요하며, 함수의 형태나 최적화 문제의 특성에 따라 다른 최적화 알고리즘을 고려할 필요가 있을 수 있습니다.

## 가우스-마르코프 정리

- 가우스-마르코프 정리(Gauss-Markov Theorem, GMT)는 특정 가정들 하에서 OLS 추정치가 최적의 선형 불편 추정치(Best Linear Unbiased Estimator, BLUE)임을 보장합니다. 즉, BLUE가 되기 위한 조건이 가우스-마르코프 정리(GMT) 이다.
- Best Linear Unbiased Estimator (BLUE)는 선형 회귀 분석에서 사용되는 중요한 통계적 개념입니다. 이 용어는 특정 가정 하에 선형 회귀 모델의 파라미터를 추정하는 최적의 방법을 설명합니다. BLUE는 다음과 같은 특징을 가진 추정치를 지칭합니다:

1. **선형성 (Linearity)**:
   추정치는 관측된 데이터의 선형 함수입니다. 이는 추정치가 데이터의 가중합으로 표현될 수 있다는 것을 의미합니다.
2. **불편성 (Unbiasedness)**:
   추정치는 편향이 없습니다. 즉, 추정치의 기대값이 실제 모델 파라미터의 값과 동일합니다. 이는 모든 가능한 표본에서 추정치가 실제 파라미터 값을 정확하게 반영하도록 보장합니다.
   $ E[\hat{\beta}] = \beta $
   여기서 $ \hat{\beta} $는 추정치, $ \beta $는 실제 파라미터 값입니다.
3. **최소 분산 (Minimum Variance)**:
   선형이고 불편한 모든 추정치 중에서, BLUE는 가장 낮은 분산을 가집니다. 이는 동일한 데이터에 대해 다른 모든 선형 불편 추정치보다 정확도가 높다는 것을 의미합니다.

- 이러한 특성은 가우스-마르코프 정리의 조건이 충족되었을 때 성립합니다. 가우스-마르코프 정리에 따르면, 회귀 모델의 오차항이 동일한 분산을 가지고 서로 독립적이며, 오차의 조건부 기대값이 0인 경우, 최소 자승법(OLS)으로 추정된 회귀 계수는 BLUE가 됩니다.
- 이는 추정치가 통계적으로 가장 신뢰할 수 있는 값임을 보장하며, 실제 데이터 분석에서 매우 중요한 성질입니다. 따라서, 이 조건들이 충족될 때 OLS 추정치를 사용하는 것은 데이터로부터 신뢰할 수 있는 정보를 얻기 위한 최선의 접근법이 됩니다.

### 가우스-마르코프 정리의 기본 가정

1. **선형성**:
   모델은 선형 관계를 유지하며, 모델 파라미터와 오차항을 포함합니다.
   $ Y = \beta_0 + \beta_1 X_1 + \dots + \beta_k X_k + \epsilon $
2. **확률적 표본**:
   입력 변수 $X_1, \ldots, X_k$는 모집단에서 무작위로 추출된 확률적 표본입니다.
3. **완전 공선성 부재**:
   입력 변수 사이에 정확한 선형 관계가 없어야 합니다. 즉, 어떠한 변수도 다른 변수의 완전한 선형 조합으로 표현될 수 없습니다.
4. **조건부 평균 오차가 0**:
   오차의 조건부 평균이 0입니다. 이는 오차가 입력 변수와 독립적임을 의미합니다.
   $ E[\epsilon | X_1, \ldots, X_k] = 0 $
5. **등분산성 (Homoscedasticity)**:
   오차는 모든 입력 변수에 대해 일정한 분산을 가져야 합니다.
   $ \text{Var}(\epsilon | X_1, \ldots, X_k) = \sigma^2 $

### 가우스-마르코프 정리의 결론

1. **효율성**: 주어진 가정 하에서, OLS 추정치는 모든 선형 불편 추정치 중에서 가장 낮은 분산을 갖습니다. 이는 OLS가 데이터의 변동성을 가장 잘 설명하며, 추정 오차를 최소화한다는 의미입니다.
1. **일관성**: 표본 크기가 증가함에 따라, OLS 추정치는 실제 모델 파라미터 값으로 수렴합니다. 이는 큰 데이터 세트에서 OLS의 효과가 더욱 명확하게 나타난다는 것을 의미합니다.

- 이 가정들이 충족되면, OLS 추정치는 최적의 추정 방법이 되며, 다른 선형 추정 방법과 비교했을 때 가장 낮은 표준 오차를 갖게 됩니다. 이는 OLS가 데이터의 기본 패턴을 효과적으로 추론할 수 있음을 의미합니다.

## 통계적 추정을 수행하는 방법

#### 핵심 용어와 정의
- **검정 통계량 (Test Statistic)**: 관심 있는 숫자에서 계산된 값으로, 일반적으로 회귀 계수가 여기에 해당합니다.
- **p-값 (p-value)**: 귀무 가설이 정확하다는 가정 하에서 검정 통계량의 관측된 값이 나타날 확률입니다.

##### 주요 가정
선형 회귀 모델에서는 다음과 같은 가정이 필요합니다:
1. **선형성**: 모델은 $ Y = X\beta + \epsilon $의 형태를 띄며, 여기서 $ \beta $는 알 수 없는 상수, $ \epsilon $은 확률 오차입니다.
2. **독립성**: 입력 변수 $ X_1, \ldots, X_k $는 모집단에서 추출된 독립 확률 표본입니다.
3. **공선성 부재**: 입력 변수 간에 정확한 선형 관계가 존재하지 않습니다.
4. **오차의 조건부 평균**: 오차의 조건부 평균은 0입니다, $ E[\epsilon | X] = 0 $.
5. **등분산성**: 오차 항의 분산은 입력 변수에 무관하게 일정합니다, $ Var(\epsilon | X) = \sigma^2 $.

#### 통계적 추론 수행
1. **회귀 계수의 추정**: 최소 자승법을 사용하여 $ \beta $의 추정치는 다음과 같습니다:
    - $ \hat{\beta} = (X^T X)^{-1} X^T Y $
1. **오차의 정규성 가정**:
    - 오차 $ \epsilon $은 정규 분포를 따릅니다. 이 가정은 신뢰 구간과 가설 검증에 사용됩니다.
1. **신뢰 구간 계산**:
    - 신뢰 구간은 통계적 추정치의 불확실성을 표현하는 범위로, 다음과 같이 계산됩니다:
    - $ \text{Confidence Interval} = \hat{\beta} \pm z \times SE(\hat{\beta}) $
    - 여기서 $ z $는 정규 분포의 특정 분위수입니다.
1. **F 통계량**:
    - 전체 회귀의 유의성을 평가하기 위해 사용되며, 추가 변수가 모델의 설명력을 얼마나 개선하는지 측정합니다.

#### 구체적 예시
- **라그랑지 승수 검증**:
  다중 제약 조건을 검증하는 데 사용되는 대안적인 방법입니다. 이는 F 검증과 유사하게 여러 모수에 대한 제한 검증을 허용합니다.

## 문제를 진단하고 해결하는 방법
- 모델 진단은 분석 결과를 정확하게 해석하고 통계적 추론을 정확히 수행하는 데 중요한 역할을 합니다. 모델이 올바르게 작동하고 있는지 확인하고, 잠재적 문제를 식별하여 잘못된 결론을 방지하기 위한 여러 절차를 포함합니다.

### 적합도
- 적합도(Goodness-of-fit)는 모델이 데이터의 변동성을 얼마나 잘 설명하는지를 평가하는 척도입니다. 이는 다양한 모델 설계를 비교하고 선택하는 데 유용하며, 모델이 데이터를 얼마나 정확하게 반영하는지를 판단할 수 있게 도와줍니다.

#### 주요 적합도 지표
1. **결정 계수 (R²)**:
   - R²은 모델이 데이터의 변동을 얼마나 잘 설명하는지를 나타내는 지표입니다. 이 값이 1에 가까울수록 모델의 설명력이 높다는 의미입니다.
   - **계산식**: $ R^2 = 1 - \frac{RSS}{TSS} $
       - TSS (Total Sum of Squares): 종속 변수 값의 총 변동성
       - RSS (Residual Sum of Squares): 잔차(실제 값과 예측 값의 차이)의 제곱 합
       - <img src="https://miro.medium.com/v2/resize:fit:1276/format:webp/0*5cX8_qVdqGHccwVg" width=600>
2. **조정된 결정 계수 (Adjusted R²)**:
   - 변수가 추가될 때마다 R²이 자동적으로 증가하는 문제를 보정합니다.
   - 추가 변수가 모델을 실질적으로 개선하지 않는 경우에는 R²의 증가를 억제합니다.
3. **정보 기준 지표**:
   - **아카이케 정보 기준 (AIC)**:
     - 모델의 복잡성에 패널티를 부과하여 우도를 조정한 값입니다. AIC가 낮을수록 모델이 데이터를 더 잘 설명한다고 평가됩니다.
     - **계산식**: $ \text{AIC} = -2 \log(L) + 2k $
   - **베이지안 정보 기준 (BIC)**:
     - AIC와 유사하지만, 표본 크기에 대한 더 큰 패널티를 부여하여 과적합을 더욱 엄격하게 통제합니다.
     - **계산식**: $ \text{BIC} = -2 \log(L) + \log(n)k $

#### 모델 선택 및 적합도 검증
- 모델 선택은 AIC와 BIC 같은 정보 기준을 통해 이루어지며, 더 낮은 값이 좋은 모델임을 나타냅니다. 실제로는 교차 검증과 같은 표본 외 검증 방법이 이상적입니다. 이러한 방법은 모델이 새로운 데이터에 대해 얼마나 잘 일반화되는지 평가하는 데 도움을 줍니다.

### 이분산성

- 이분산성은 선형 회귀 모델에서 잔차의 분산이 입력 변수에 따라 일정하지 않고 변화하는 현상을 말합니다.
- 이는 평균에서 먼 입력값에 대해 오차가 커지는 경향이 있는 경우 흔히 발생합니다.
- 이로 인해 OLS(최소 제곱 추정법)의 표준 오차 추정치가 실제보다 낮게 나와, 잘못된 통계적 발견을 유발할 수 있습니다.
- <img src="https://blog.kakaocdn.net/dn/8OpHV/btqWXalQTEd/DzPhFNrqyG3oWZZ9Zk6jpK/img.png" width=600>

#### 이분산성의 진단

- **시각적 검사**: 잔차의 체계적인 패턴을 육안으로 검사하여 이분산성의 존재 여부를 추정합니다.
- **통계적 검증**:
  - **브로이슈-파간(Breusch-Pagan) 검정**: 잔차 분산의 동질성을 검증하는 방법입니다.
  - **화이트(White) 검정**: 마찬가지로 이분산성을 검증하는 데 사용됩니다.

#### 이분산성에 대한 대응 방법
1. **조정된 표준 오차 사용**:
   - 화이트 표준 오차와 같은 샌드위치 추정치를 사용하여 이분산성을 고려한 오차 분산을 계산합니다.
   - 일반 표준 오차가 모든 데이터 포인트의 분산이 동일하다고 가정할 때, 조정된 표준 오차는 데이터의 불균등한 분산이나 데이터 포인트 간의 상관관계 같은 현실적 요소를 반영합니다. 
2. **클러스터링된 표준 오차**:
   - 데이터가 동질적으로 구별되는 그룹별로 다른 오차 분산을 가정합니다.
   - 이 방법은 특히 데이터가 특정 업종이나 자산 클래스로 구분될 때 유용합니다.
3. **가중 최소 자승법 (WLS)**:
   - 공분산 행렬이 이분산 오차를 포함하는 경우, WLS는 각 데이터 포인트의 중요성을 가중치로 조정하여 추정을 수행합니다.
4. **일반화 최소 자승법 (GLS)**:
   - 임의 공분산 행렬 구조를 가정하며, 이분산성 또는 계열 상관이 있는 경우 효율적이고 불편한 추정치를 제공합니다.

### 계열 상관

- 계열 상관(Serial Correlation)은 연속적인 데이터 포인트 간의 잔차가 상관관계를 보이는 현상을 말합니다.
- 이는 선형 회귀 모델에서 중요한 문제가 될 수 있으며, 다음과 같은 방법으로 진단 및 해결할 수 있습니다.
- <img src="https://velog.velcdn.com/images/soonwoo2003/post/96e15078-ebdc-45d6-9cc0-ce45d1a7c812/image.png">

- **계열 상관**과 **자기 상관**은 모두 시계열 데이터 분석에서 사용되는 통계적 도구입니다만, 그들이 측정하는 관계의 성격이 다릅니다.
    - 1. **계열 상관 (Cross-correlation)**:
       - 계열 상관은 두 시계열 데이터 세트 간의 관계를 측정합니다. 예를 들어, 한 시계열이 다른 시계열을 어떻게 예측할 수 있는지, 또는 두 데이터 세트 사이에 어떤 시차(time lag)가 존재하는지를 파악할 때 사용됩니다.
       - 이는 두 다른 변수 간의 상관 관계를 분석하는 데 사용되며, 예를 들어 주가 지수와 특정 산업의 주식 가격 사이의 관계 등을 분석할 때 활용됩니다.
    
    - 2. **자기 상관 (Autocorrelation)**:
       - 자기 상관은 하나의 시계열 데이터 내에서, 시간의 흐름에 따라 데이터 포인트들 간의 상관 관계를 측정합니다. 이는 시계열 데이터가 시간에 따라 얼마나 일관된 패턴을 보이는지, 예를 들어 일정 시간 간격으로 반복되는 패턴이 있는지 등을 파악할 때 유용합니다.
       - 주로 계절성 패턴이나 주기적 변동성을 파악하거나, 예측 모델을 개발하는 데 사용됩니다.

#### 진단 방법:
- **더빈-왓슨(Durbin-Watson) 통계량**을 사용하여 계열 상관을 진단합니다. 이 통계량은 0에서 4 사이의 값을 가지며, 값이 2에 가까울 경우 잔차 간에 자기 상관이 없다고 간주합니다.
  - 값이 2보다 작으면 양의 자기 상관을,
  - 값이 2보다 크면 음의 자기 상관을 나타냅니다.
#### 계열 상관의 문제점:
- 양의 계열 상관은 표준 오차를 과소 평가하여 통계적으로 유의하지 않은 결과를 유의하다고 잘못 판단할 수 있습니다.
- 이는 잘못된 발견(false discoveries)으로 이어질 수 있습니다.
#### 문제 해결:
- 계열 상관을 고려한 **조정된 표준 오차**를 계산하여 모델의 통계량을 보정합니다.
- 정확한 임곗값을 설정하기 위해서는 모수 및 관측치 수에 따라 통계적 표를 참조해야 합니다.

### 다중 공선성
- 다중 공선성은 회귀 분석에서 두 개 이상의 설명 변수(독립 변수) 사이에 높은 상관관계가 있을 때 발생합니다.
- 이 현상은 회귀 모델의 정확성과 신뢰성에 몇 가지 문제를 일으킬 수 있습니다.
#### 문제점
1. **팩터 결정 어려움**: 서로 상관관계가 높은 변수들 때문에 각 변수가 종속 변수에 미치는 영향을 분명히 파악하기 어렵습니다.
2. **p-값의 오해**: 실제로 중요한 변수라도 p-값이 높게 나타나 오해의 여지가 있습니다. 이는 변수가 통계적으로 유의미하지 않은 것처럼 보이게 할 수 있습니다.

#### 해결 방법
- **변수 제거**: 상관관계가 높은 변수 중 하나를 제거하여 모델을 단순화시킬 수 있습니다.
- **표본 크기 증가**: 데이터의 양을 늘리면 각 변수의 독립적인 영향력을 더 정확하게 추정할 수 있습니다.

## statsmodels을 활용한 최소 자승법

```python
X = sm.add_constant(data['X'])
model = sm.OLS(data['Y'], X).fit()
print(model.summary())
```

- 오른쪽 상단 R-square, F 통계량
- t통계량은 추정값을 표준 오차로 나눈 값으로, t값이 커서 절편과 기울기 계수가 매우 유의미 하다고 판단 (t값이 클수록 p값은 작아짐)
- 하단에는 잔차 진단, 왜도와 첨도가 포함
- 옴니버스와 자크-베라 검정은 잔차가 정규 분포한다는 귀무가설을 기각하지 못한다.(둘다 p값으로 0.05 이하 일 때 귀무 가설 기각하고 잔차가 정규 분포를 따르지 않는다고 결론 내릴수 있다.)
- 더빈 왓슨은 잔차들 간의 자기 상관을 검증 하는 것으로, 귀무가설은 '잔차들 간에 자기상관이 없다' 이며, 0 부터 4 까지의 값을 가진다. 2 보다 작으면 양의 자기 상관, 크면 음의 자기 상관, 1.5미만, 2.5 초과 일때 자기 상관이 있다고 보고 귀무가설 기각. 위 값은 이에 해당하지 않으면 기각 할수 없어, 자기 상관이 없다고 결론.

## CAPM에서 파마-프렌치 5 팩터 모델까지

- CAPM (자본 자산 가격 결정 모델, Capital Asset Pricing Model)은 금융에서 자산의 기대 수익률을 예측하는 모델로, 투자의 위험성을 평가하는 데 주로 사용됩니다.
- 이 모델은 금융 자산의 수익률이 시장 전체의 위험에 어떻게 반응하는지를 설명하며, 이를 통해 자산의 적절한 요구 수익률을 계산할 수 있습니다.

### CAPM 공식:
- $ E(R_i) = R_f + \beta_i (E(R_m) - R_f) $
    - $ E(R_i) $는 자산 $ i $의 기대 수익률입니다.
    - $ R_f $는 무위험 수익률로, 국채 같은 위험이 없는 투자에서 기대할 수 있는 수익률을 의미합니다.
    - $ \beta_i $는 자산의 시장 위험 대비 민감도를 나타내며, 시장 포트폴리오 $ E(R_m) $ 대비 자산의 수익률 변동성을 측정합니다.
        - 1. **수익률 데이터 수집**: 분석하고자 하는 자산과 시장 지수(예: S&P 500)의 과거 수익률 데이터를 수집합니다.
        - 2. **수익률 변화 계산**: 수집한 데이터를 바탕으로 각 기간에 대한 자산과 시장 지수의 수익률 변화를 계산합니다.
        - 3. **공분산 계산**: 자산 수익률 변화와 시장 지수 수익률 변화의 공분산을 계산합니다.
        - 4. **분산 계산**: 시장 지수 수익률 변화의 분산을 계산합니다.
        - 5. **베타 계산**: 공분산을 분산으로 나누어 베타값을 산출합니다.
        - $ \beta = \frac{\text{Cov}(R_i, R_m)}{\text{Var}(R_m)} $
    - $ E(R_m) $는 시장 포트폴리오의 기대 수익률입니다.
    - $ E(R_m) - R_f $는 시장 위험 프리미엄이며, 시장이 무위험 수익률보다 추가로 제공하는 기대 수익률을 나타냅니다.
 

### CAPM의 사용:
- CAPM은 개별 자산이나 포트폴리오의 위험을 고려한 적정 수익률을 산출하기 위해 사용됩니다.
- 투자자는 이 모델을 사용하여 투자 결정을 내리고, 과도하게 위험한 자산에 대해 더 높은 수익을 요구할 수 있습니다.

### 이해하기 쉬운 예:
- 예를 들어, 어떤 주식의 $ \beta $가 1.5라면, 시장 수익률이 1% 상승할 때 해당 주식의 수익률은 평균적으로 1.5% 상승할 것으로 예상됩니다.
- 만약 무위험 수익률이 2%이고, 시장의 기대 수익률이 8%라면, 이 주식의 기대 수익률은 다음과 같이 계산됩니다:
- $ E(R_i) = 2\% + 1.5 \times (8\% - 2\%) = 11\% $
- 이 모델은 투자의 위험을 수치화하고, 투자자가 위험을 감안한 적정 수익을 요구할 수 있게 해줍니다.

### CAPM의 한계와 발전:
- **이례 현상의 발견**: 소규모 기업의 높은 수익률 등 CAPM으로 설명할 수 없는 다양한 리스크 조정 수익률이 관찰됩니다.
- **다중 팩터 모델의 등장**: CAPM의 한계를 극복하기 위해, 여러 리스크 팩터를 고려하는 다중 팩터 모델이 개발되었습니다.

### 파마-맥베스 회귀 분석의 두 단계:

1. **첫 번째 단계: 팩터 적재량 추정**
   - 각 자산이나 포트폴리오에 대한 시계열 회귀를 수행하여 팩터에 대한 노출을 추정합니다. 이 과정은 각 자산에 대해 다음 수식을 사용하여 수행됩니다.
     $ F_{tx1} = T \times (M+1) + (M+1) \times 1 + T \times 1 $
   - 여기서 $ F_{tx1} $은 초과 수익률, $ T $는 시간, $ M $은 팩터의 개수입니다.

2. **두 번째 단계: 리스크 프리미엄 추정**
   - 각 기간마다 수행되는 횡단면 회귀 분석을 통해 리스크 프리미엄의 벡터를 추정합니다. 이 과정은 다음 수식을 사용하여 수행됩니다.
     $ N \times (M+1) = N \times (M+1) + (M+1) \times 1 $
   - 각 기간 동안 얻은 리스크 프리미엄 벡터를 시간 평균으로 계산하여 리스크 프리미엄 추정치를 얻습니다.

- 이 방법론은 특히 주식 시장 데이터에 대해 리스크 팩터와 수익률의 관계를 효과적으로 분석하는 데 사용됩니다. 이 과정을 통해 투자자와 금융 분석가는 특정 리스크 요인에 대한 포트폴리오의 민감도를 이해하고 적절한 투자 결정을 내릴 수 있습니다.

```python
data_path = '/home/restful3/datasets_bigpond/ml4t/assets.h5'
```

```python
with pd.HDFStore(f'{data_path}') as store:
    # 'prices' 테이블에서 2010년부터 2017년까지의 조정 종가를 로드하고 피벗하여 테이블 형태로 재구성합니다.
    prices = store['/quandl/wiki/prices'].adj_close.unstack().loc['2010':'2017']
    # 'stocks' 테이블에서 중복을 제거한 주식 데이터를 로드합니다.    
    equities = store['/us_equities/stocks'].drop_duplicates()
```

```python
# 섹터 정보를 이용하여 각 주식의 섹터를 딕셔너리 형태로 저장합니다.
sectors = equities.filter(prices.columns, axis=0).sector.to_dict()
# 섹터 정보를 바탕으로 주식 데이터를 필터링하고 결측치가 있는 열을 제거합니다.
prices = prices.filter(sectors.keys()).dropna(how='all', axis=1)
```

```python
# 가격 데이터로부터 월별 수익률을 계산합니다.
returns = prices.resample('M').last().pct_change().mul(100).to_period('M')
returns = returns.dropna(how='all').dropna(axis=1)
returns.info()
```

```python
# Fama-French 모델 데이터를 'returns' 인덱스에 맞춰서 필터링합니다.
ff_factor_data = ff_factor_data.loc[returns.index]
ff_portfolio_data = ff_portfolio_data.loc[returns.index]
```

```python
ff_factor_data.describe()
```

```python
# 리스크 프리 수익률을 초과 수익률에서 제거합니다.
excess_returns = returns.sub(ff_factor_data.RF, axis=0)
excess_returns.info()
```

```python
# 초과 수익률의 1백분위수와 99백분위수 사이로 값을 제한합니다.
excess_returns = excess_returns.clip(lower=np.percentile(excess_returns, 1),
                                     upper=np.percentile(excess_returns, 99))
```

```python
ff_portfolio_data.info()
```

- **첫 번째 단계 - 팩터 적재량 추정**: 여기서는 각 산업의 포트폴리오에 대한 Fama-French 팩터 데이터를 사용하여 회귀분석을 수행합니다. 이 과정은 각 포트폴리오의 팩터에 대한 노출(베타 계수)을 추정하는 데 사용됩니다. `OLS` 회귀분석을 통해 각 포트폴리오의 팩터 노출을 계산합니다.

```python
# 각 산업별 베타 계수를 계산합니다.
betas = []
for industry in ff_portfolio_data:
    # Fama-French 요인 데이터와 산업 데이터를 사용하여 OLS 회귀분석을 수행합니다.    
    step1 = OLS(endog=ff_portfolio_data.loc[ff_factor_data.index, industry], 
                exog=add_constant(ff_factor_data)).fit()
    # 회귀분석 결과에서 상수항을 제외한 파라미터들을 저장합니다.    
    betas.append(step1.params.drop('const'))
```

```python
betas = pd.DataFrame(betas, 
                     columns=ff_factor_data.columns, 
                     index=ff_portfolio_data.columns)
betas.info()
```

- **두 번째 단계 - 리스크 프리미엄 추정**: 추정된 베타 계수를 사용하여 각 기간에 대한 리스크 프리미엄을 추정합니다. 이 단계에서는 모든 포트폴리오의 베타 계수를 독립 변수로 하고, 해당 기간의 포트폴리오 수익률을 종속 변수로 사용하는 횡단면 회귀분석을 수행합니다.

```python
# 각 시간 단위에 대해 람다(리스크 프리미엄)를 계산합니다.
lambdas = []
for period in ff_portfolio_data.index:
    # 이전 단계에서 구한 베타 계수를 사용하여 OLS 회귀분석을 수행합니다.    
    step2 = OLS(endog=ff_portfolio_data.loc[period, betas.index], 
                exog=betas).fit()
    # 회귀분석 결과 파라미터를 저장합니다.    
    lambdas.append(step2.params)
```

```python
lambdas = pd.DataFrame(lambdas, 
                       index=ff_portfolio_data.index,
                       columns=betas.columns.tolist())
lambdas.info()
```

```python
lambdas.mean()
```

```python
# 람다의 평균과 표준편차를 사용하여 t-통계량을 계산합니다.
t = lambdas.mean().div(lambdas.std())
t
```

# 수축 방법을 사용한 선형 회귀 규제화

- 선형 회귀 모델에 적용된 수축 방법은 모델의 예측 성능을 향상시키고, 과적합을 방지하는 데 중요한 역할을 합니다.
  
#### 기본 개념
- **선형 회귀의 전통적 방법**: 최소제곱법(OLS)은 가우스-마르코프 가정 하에 최적의 편향 없는 추정치를 제공합니다.
- **오차 공분산 행렬의 가정 위반**: OLS의 기본 가정이 위반되더라도 일반화 최소제곱법(GLS) 같은 방법이 유사하게 효과적입니다.

#### 문제점
- **상관 변수의 포함**: 선형 회귀 모델에 많은 상관 변수가 포함되면, 계수가 제대로 결정되지 않을 위험이 있습니다.
- **분산 증가의 리스크**: 상관 변수 간의 상쇄 효과로 인해 모델이 과도하게 표본 데이터에 적합해질 수 있으며, 이는 높은 분산과 예측 오류를 증가시킬 수 있습니다.

#### 수축 방법의 필요성
- **분산 감소 및 일반화 오류 최소화**: 수축 방법은 모델의 계수에 제약을 추가하여 모델이 데이터에 과도하게 적합되는 것을 방지하고, 일반화 오류를 줄이는데 도움을 줍니다.
- **규제화 기법**: 계수에 대한 페널티를 추가하여 모델의 복잡성을 관리하고, 더 높은 예측 정확성을 달성하기 위해 사용됩니다.

## 과적합을 방지하는 방법

#### **규제화 개요**
- 규제화는 계수의 크기를 제한하여 모델의 복잡성을 줄이고, 과대적합을 방지합니다. 이 방법은 모델의 예측력을 향상시킬 수 있으며, 일반적인 오차 함수에 페널티 항을 추가하여 실행됩니다.

#### **수축 방법**
- 수축 방법은 모델의 예측 정확도를 높이고 해석을 단순화하기 위해 사용됩니다. 주로 두 가지 목적을 가집니다:
   - **예측 정확도 향상**: 수축 방법은 계수를 제로 또는 제로에 가깝게 만들어 모델의 분산을 줄이고, 예측 오류의 리스크를 감소시킵니다. 이는 일부 계수를 0으로 설정하여 일반화 오류를 줄이는 데 도움이 됩니다.
   - **모델 해석의 단순화**: 많은 예측 변수가 포함된 모델은 해석이 복잡해질 수 있습니다. 수축 방법은 가장 강한 효과를 가진 변수만을 모델에 포함시켜 나머지는 제외함으로써 모델을 단순화합니다.

#### **수축 모델의 구현**
- 수축 모델은 목표 함수에 페널티 항 $S(β)$을 추가하여 구현됩니다. 이 항은 RSS(잔차 제곱합)와 계수의 크기에 대한 페널티를 포함합니다. 페널티는 보통 라그랑지안 형태로 최소화 문제에 포함됩니다:
   - $L(β) = RSS + λ * S(β)$
   - 여기서 $λ$는 규제화 파라미터로, 규제의 강도를 결정하며, 이 값이 커질수록 계수에 대한 제약이 강해집니다.

#### **주요 규제화 모델**
   - **리지 회귀 (Ridge Regression)**: 계수의 제곱의 합을 최소화하여 계수의 크기를 줄입니다.
   - **라쏘 회귀 (Lasso Regression)**: 계수의 절대값의 합을 최소화하여 일부 계수를 정확히 0으로 만듭니다. 이는 변수 선택 효과가 있어 모델의 해석을 더욱 단순화합니다.
   - **엘라스틱 넷 (Elastic Net)**: 리지와 라쏘의 특징을 결합한 모델로, 두 가지 페널티를 모두 사용합니다.

#### **페널티 선택과 모델 최적화**
- 실제 모델 훈련에서는 교차 검증을 통해 최적의 $λ$ 값을 선택하게 됩니다. 이 값은 모델이 데이터에 과대적합하지 않으면서 최대의 예측 정확도를 달성하도록 도와줍니다. 

- 이러한 규제화 방법들은 모델의 일반화 능력을 향상시키고, 훈련 데이터에 대한 과적합을 방지하는 데 중요한 역할을 합니다.

## 리지 회귀 분석의 작동 방식

#### **기본 개념**
- 리지 회귀는 선형 회귀 모델에 L2 규제를 적용한 것으로, 계수의 제곱합을 페널티로 추가합니다.
- 이 방식은 모델의 과대적합을 줄이고 계수를 수축시키는 데 도움을 줍니다.

#### **리지 회귀의 정의**
- 리지 회귀의 목표 함수는 다음과 같이 정의됩니다:
- $ \text{Ridge} = \arg \min_{\beta} \left( (y - X\beta)^T (y - X\beta) + \lambda \|\beta\|^2 \right) $
- 여기서 $ \lambda $는 규제 파라미터로, 계수의 크기를 조절합니다.

#### **절편의 처리**
- 리지 회귀에서 절편은 페널티에서 제외되어, 모델이 데이터의 중심을 기준으로 계수를 조정할 수 있게 합니다. 이는 모든 출력값에 동일한 상수를 추가하더라도 모델의 기울기 파라미터에 영향을 주지 않도록 합니다.

#### **데이터의 표준화**
- 입력 변수를 표준화하는 것이 중요합니다. 이는 리지 회귀가 입력 변수의 스케일에 민감하기 때문입니다. 각 입력 변수에서 평균을 빼고 표준 편차로 나누어줌으로써 이루어집니다.

#### **리지 회귀의 솔루션**
- 리지 회귀의 솔루션은 닫힌 형태로 주어집니다:
- $ \beta_{\text{Ridge}} = (X^T X + \lambda I)^{-1} X^T y $
- 이 식은 $ X^T X $가 전체 순위가 아닐 때도 단일이 아니게 하는 $ \lambda I $를 통해 보정합니다. 이는 계수에 대한 L2 규제를 나타냅니다.

#### **직교 정규 입력의 특성**
- 직교 정규 입력이 있는 경우, 리지 추정치는 최소제곱 추정치의 축소된 버전으로 나타나며, 각 계수는 다음과 같이 수축됩니다:
- $ \beta_{\text{Ridge}} = \frac{\beta_{\text{LS}}}{1 + \lambda} $
- 여기서 $ \beta_{\text{LS}} $는 최소제곱 추정치를 나타냅니다.

#### **SVD와 리지 회귀의 상관관계**
- SVD(특이값 분해)를 사용하여 입력 행렬 $ X $를 분석함으로써, 리지 회귀가 입력의 다양한 방향에 따라 계수를 어떻게 수축시키는지 이해할 수 있습니다. 이는 데이터의 주요 구성 요소를 기반으로 계수의 수축을 설명합니다.

- 이러한 리지 회귀의 설명은 분산이 높은 데이터 방향이 출력 예측에 더 중요하다는 암묵적 가정을 가지고 있습니다. 이 방법은 회귀 계수를 통제하고 모델의 예측 정확도를 향상시키는 데 유용합니다.

## 라쏘 회귀 분석의 작동 방식

#### 기본 개념
- 라쏘 회귀는 신호 처리와 통계학에서 널리 사용되는 방법으로, 회귀 계수의 절댓값의 합에 대한 페널티를 추가하여 계수를 수축시킵니다. 이 페널티는 L1 규제라고 하며, 계수를 완전히 0으로 만들 수 있어 특성 선택에 유용합니다.

#### 수식
- 라쏘 회귀의 목적 함수는 다음과 같이 정의됩니다:
- $ \text{B}_{\text{Lasso}} = \arg\min \left( (y - XB)^2 + \lambda \sum |B| \right) $
- 여기서 $ \lambda $는 페널티의 강도를 조절하는 규제화 파라미터이며, $ B $는 계수 벡터입니다.

#### 특징
- **표준화**: 라쏘 회귀를 사용하기 전에 입력 변수를 표준화하는 것이 중요합니다. 이는 모든 변수가 동일한 스케일을 갖도록 보장하여, 라쏘 모델이 변수 크기에 영향을 받지 않도록 합니다.
- **비선형 솔루션**: 라쏘 페널티는 비선형 솔루션을 만들며, 계수에 대한 폐쇄형 해가 존재하지 않습니다. 따라서, 수치적 최적화 방법을 통해 계수를 추정합니다.
- **계수 경로**: 라쏘는 규제화 강도 $ \lambda $의 다양한 값에 대해 계수의 경로를 계산하는 효율적인 알고리즘을 사용합니다. 이를 통해 다양한 규제화 수준에서의 계수 값을 탐색할 수 있습니다.
- **특성 선택**: 라쏘는 일부 계수를 정확히 0으로 만드는 특성을 가지므로, 불필요한 변수를 제거하고 모델을 단순화하는 데 도움을 줍니다.

#### 실용화
- 라쏘 회귀는 다양한 선형 회귀 모델을 실용화하고 주식 거래 신호와 같은 예측 가능한 결과를 생성하는 데 활용됩니다. 이는 모델이 데이터의 주요 신호를 포착하면서도 과적합을 피할 수 있도록 도와줍니다.

- 볼리저 밴드 계산

```python
from talib import BBANDS
import pandas as pd
from tqdm import tqdm

# 볼린저 밴드 계산 함수 정의
def compute_bb(close):
    high, mid, low = BBANDS(close, timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)
    return pd.DataFrame({'bb_high': high, 'bb_low': low}, index=close.index)

# 결과를 저장할 빈 DataFrame 준비
bb_high_series = pd.Series(index=prices.index)
bb_low_series = pd.Series(index=prices.index)

# 각 ticker에 대해 볼린저 밴드 계산
for ticker in tqdm(prices.index.get_level_values('ticker').unique()):
    # 현재 ticker의 close 가격 데이터 선택
    ticker_data = prices.loc[(slice(None), ticker), 'close']
    
    # 볼린저 밴드 계산
    bb_data = compute_bb(ticker_data)
    
    # 계산된 볼린저 밴드 값을 저장할 Series에 추가
    bb_high_series.loc[(slice(None), ticker)] = bb_data['bb_high']
    bb_low_series.loc[(slice(None), ticker)] = bb_data['bb_low']

# 계산된 볼린저 밴드 값을 prices DataFrame에 추가
prices['bb_high'] = bb_high_series
prices['bb_low'] = bb_low_series
```

- ATR(Average True Range) : 고가, 저가, 종가를 취하는 지표

```python
import talib
import pandas as pd
from tqdm import tqdm

# ATR 계산 함수 정의
def compute_atr(stock_data):
    # talib ATR 함수를 사용하여 ATR 계산
    atr = talib.ATR(stock_data['high'], stock_data['low'], stock_data['close'], timeperiod=14)
    # 계산된 ATR의 평균과 표준편차로 정규화
    normalized_atr = (atr - atr.mean()) / atr.std()
    return normalized_atr

# 결과를 저장할 빈 Series 준비
atr_series = pd.Series(index=prices.index)

# 각 ticker에 대해 ATR 계산
for ticker in tqdm(prices.index.get_level_values('ticker').unique()):
    # 현재 ticker의 데이터 선택
    ticker_data = prices.loc[(slice(None), ticker), ['high', 'low', 'close']]
    
    # ATR 계산
    atr_values = compute_atr(ticker_data)
    
    # 계산된 ATR 값을 저장할 Series에 추가
    atr_series.loc[(slice(None), ticker)] = atr_values

# 계산된 ATR 값을 prices DataFrame에 추가
prices['atr'] = atr_series
```

- MACD 지표
    - 짧은 기간과 자기 지수 이동평균 간의 차이

```python
import talib
import numpy as np
import pandas as pd
from tqdm import tqdm

# MACD 계산 및 정규화 함수 정의
def compute_macd(close):
    # MACD 계산
    macd, macdsignal, macdhist = talib.MACD(close, fastperiod=12, slowperiod=26, signalperiod=9)
    # 계산된 MACD의 평균과 표준편차로 정규화
    normalized_macd = (macd - np.mean(macd)) / np.std(macd)
    return normalized_macd

# 결과를 저장할 빈 Series 준비
macd_series = pd.Series(index=prices.index)

# 각 ticker에 대해 MACD 계산
for ticker in tqdm(prices.index.get_level_values('ticker').unique()):
    # 현재 ticker의 종가 데이터 선택
    close_prices = prices.loc[(slice(None), ticker), 'close']
    
    # MACD 계산
    macd_values = compute_macd(close_prices)
    
    # 계산된 MACD 값을 저장할 Series에 추가
    macd_series.loc[(slice(None), ticker)] = macd_values

# 계산된 MACD 값을 prices DataFrame에 추가
prices['macd'] = macd_series
```

### 시차 수익률 추가
- 다양한 과거 시차에 대한 가격 추세 포착

```python
import pandas as pd
from tqdm import tqdm

q = 0.0001
lags = [1, 5, 10, 21, 42, 63]

# Lag 별로 수익률을 계산하는 함수 정의
def compute_returns(data, lags):
    for lag in lags:
        # 시차에 따른 백분율 변화 계산
        returns = data['close'].pct_change(lag)
        
        # 하위 및 상위 백분위수로 값을 제한
        lower_quantile = returns.quantile(0.0001)
        upper_quantile = returns.quantile(0.9999)
        returns = returns.clip(lower=lower_quantile, upper=upper_quantile)
        
        # 연간화 수익률 계산
        returns = (1 + returns).pow(1 / lag) - 1
        
        # 새로운 컬럼 이름 설정
        column_name = f'return_{lag}d'
        
        # 계산된 수익률을 데이터프레임에 추가
        data[column_name] = returns
        
    return data

# 결과 DataFrame 준비
result_prices = pd.DataFrame()

# 각 ticker에 대해 수익률 계산
for ticker in tqdm(prices.index.get_level_values('ticker').unique()):
    # 현재 ticker의 데이터 선택
    ticker_data = prices.loc[(slice(None), ticker), :]
    
    # 수익률 계산
    ticker_data = compute_returns(ticker_data, lags)
    
    # 계산된 데이터를 결과 DataFrame에 추가
    result_prices = pd.concat([result_prices, ticker_data])

# 결과를 원본 prices DataFrame에 병합
prices = result_prices.copy()
```

1. Omnibus (오브니버스):
   - Omnibus는 정규성 검정 통계량입니다. 이 값이 클수록 모델의 잔차가 정규분포를 따른다는 가정에 부합합니다. 여기서 Omnibus 값이 매우 크므로 모델의 잔차가 정규분포를 따르지 않는다는 증거가 됩니다.

2. Prob(Omnibus):
   - 이 값은 Omnibus 검정의 p-value를 나타냅니다. 여기서 p-value가 0.000으로 매우 작으므로, 모델의 잔차가 정규분포를 따르는 것이 아니라고 할 수 있습니다.

3. Jarque-Bera (JB):
   - Jarque-Bera 검정 통계량은 잔차의 왜도와 첨도를 기반으로 하며, 정규성을 검정합니다. JB 값이 매우 크므로 잔차의 정규성 가정이 만족되지 않음을 시사합니다.

4. Prob(JB):
   - 이 값은 Jarque-Bera 검정의 p-value를 나타냅니다. 여기서도 p-value가 0.00으로 매우 작으므로, 잔차의 정규성 가정이 만족되지 않는다는 것을 나타냅니다.

5. Skew (왜도):
   - 왜도는 분포의 비대칭 정도를 나타내며, 여기서는 -0.250으로 대부분의 경우에 대칭에 가까운 것으로 간주됩니다.

6. Kurtosis (첨도):
   - 첨도는 분포의 뾰족한 정도를 나타내며, 14.767로 매우 뾰족한 분포임을 나타냅니다. 뾰족한 정도가 높을수록 꼬리 부분에 대량의 데이터가 존재하고, 분포가 중심 부분에 집중되어 있음을 의미합니다.

7. Durbin-Watson:
   - Durbin-Watson 통계량은 잔차의 자기상관을 검정합니다. 이 값이 0에 가까울수록 양의 자기상관을 가지며, 2에 가까울수록 자기상관이 없다고 판단됩니다. 여기서는 0.437로 양의 자기상관이 있는 것으로 보입니다.

8. Cond. No. (Condition Number):
   - Condition Number는 다중공선성의 정도를 나타냅니다. 일반적으로 100보다 크면 다중공선성이 발생할 가능성이 있습니다. 여기서는 1.33e+12로 매우 큰 값이므로 다중공선성이 있을 가능성이 있습니다.

## 사이킷런을 이용한 리지 회귀 분석

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import cross_val_score

# alpha 값 범위 설정
alphas = np.logspace(-4, 4, 100)

# Ridge 모델 초기화 (RidgeCV를 사용하여 교차 검증을 수행)
ridge_cv = RidgeCV(alphas=alphas, cv=5)  # cv는 교차 검증의 fold 수를 지정합니다.

# 모델 훈련
ridge_cv.fit(X, y)

# 최적의 alpha 값 확인
best_alpha = ridge_cv.alpha_
print("최적의 alpha 값:", best_alpha)

# alpha 값에 따른 Ridge 모델의 계수 변화 기록
coefficients = []

for alpha in tqdm(alphas):
    ridge_model = Ridge(alpha=alpha)
    ridge_model.fit(X, y)
    coefficients.append(ridge_model.coef_)
```

```python
# 그래프로 표시
plt.figure(figsize=(10, 6))

for i in range(len(X.columns)):
    plt.plot(alphas, [coef[i] for coef in coefficients], label=X.columns[i])

plt.xscale('log')
plt.xlabel('alpha')
plt.ylabel('Coefficient')
plt.title('Ridge Coefficients vs. Alpha')
plt.legend(loc='upper right', fontsize='small')  # 범례 위치와 크기 조정
plt.show()
```

## 로지스틱 회귀 모델

- 로지스틱 회귀 모델은 선형 회귀 모델과 유사하게 독립 변수 $ x $의 선형 함수를 사용하여 종속 변수 $ y $의 확률을 모델링합니다. 이 모델은 주어진 입력 변수 $ x $가 주어졌을 때 해당 클래스에 속할 확률을 추정합니다. 이를 통해 이진 분류 문제를 해결할 수 있습니다.

### 로지스틱 함수
- 로지스틱 회귀 모델은 로지스틱 함수를 사용하여 주어진 입력 변수 $ x $에 대한 출력을 확률로 변환합니다. 로지스틱 함수는 다음과 같은 수식으로 표현됩니다.
- $p(x) = \frac{1}{1 + e^{-z}}$
- 여기서 $ z $는 입력 변수 $ x $의 선형 조합이고, $ e $는 자연 상수입니다. 이 함수는 항상 S 모양의 곡선을 생성하며, 입력 변수 $ x $의 값에 관계없이 0과 1 사이의 출력을 생성합니다.

### 로지스틱 회귀 모델
- 로지스틱 회귀 모델은 입력 변수 $ x $의 선형 조합에 로지스틱 함수를 적용하여 종속 변수 $ y $의 확률을 예측합니다. 이 모델은 선형 회귀와 유사하지만, 출력이 확률로 변환되므로 종속 변수가 이진 형태인 경우에 사용됩니다.

### 최대 우도 추정
- 로지스틱 회귀 모델의 계수는 최대 우도 추정을 사용하여 추정됩니다. 이는 주어진 훈련 데이터에서 관측 된 결과와 모델의 예측 간의 유사성을 최대화하는 방법입니다. 이를 통해 모델이 주어진 입력 변수에 대해 올바른 예측을 할 수 있도록 합니다.

## statsmodels로 추론을 수행하는 방법

```python
# statsmodels 라이브러리에서 미국 매크로 경제 데이터를 불러옵니다.
data = pd.DataFrame(sm.datasets.macrodata.load().data)
data.info()
```

```python
# 실질 GDP의 연간 성장률을 계산하고 새 열에 저장합니다.
data['growth_rate'] = data.realgdp.pct_change(4)

# 성장률이 20기간 이동 평균을 넘는지 여부를 나타내는 이진 목표 변수를 만듭니다.
# 결과를 -1만큼 이동하여 예측에 적합한 선행 지표로 만듭니다.
data['target'] = (data.growth_rate > data.growth_rate.rolling(20).mean()).astype(int).shift(-1)

# 숫자 분석을 위해 'quarter' 열 데이터 유형을 정수로 변환합니다.
data.quarter = data.quarter.astype(int)
```

```python
# 백분율 변경을 계산할 열의 목록을 지정합니다.
pct_cols = ['realcons', 'realinv', 'realgovt', 'realdpi', 'm1']

# 모델에 필요하지 않은 열을 삭제하는 열의 목록입니다.
drop_cols = ['year', 'realgdp', 'pop', 'cpi', 'growth_rate']

# 선택한 열에 대한 백분율 변경을 계산하고 업데이트합니다.
data.loc[:, pct_cols] = data.loc[:, pct_cols].pct_change(4)
```

```python
# 모델링을 위해 데이터를 준비합니다: 불필요한 열 삭제, 'quarter'를 더미 변수(원-핫 인코딩)로 인코딩하고
# 더미 변수 함정을 피하기 위해 첫 번째 분기를 제거합니다.
data = pd.get_dummies(data.drop(drop_cols, axis=1), columns=['quarter'], drop_first=True).dropna()
```

## 가격 움직임을 로지스틱 회귀 분석으로 예측

### 회귀 문제를 분류 문제로 변환하는 방법

- 수익율이 양수면 1 아니면 0

```python
y = (y >0).astype(int)
```

### 로지스틱 회귀 하이퍼파라미터 교차 검증

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score
from scipy.stats import spearmanr
from sklearn.model_selection import TimeSeriesSplit

# 가정: X와 y 데이터프레임이 이미 정의되어 있으며, y는 단일 값으로 주어짐

# 교차 검증 설정
n_splits = 4 * 252
cv = TimeSeriesSplit(n_splits=n_splits, max_train_size=252)

# C 값에 대한 로그 공간 설정
Cs = np.logspace(-5, 5, 11)

# 모델 훈련과 평가
results = []
for C in Cs:
    model = LogisticRegression(C=C, fit_intercept=True, solver='liblinear')  # liblinear 추천
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])
    
    for i, (train_idx, test_idx) in enumerate(cv.split(X), 1):
        X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
        X_test, y_test = X.iloc[test_idx], y.iloc[test_idx]
        pipe.fit(X=X_train, y=y_train)
        y_score = pipe.predict_proba(X_test)[:, 1]
        auc = roc_auc_score(y_true=y_test, y_score=y_score)
        ic, pval = spearmanr(y_score, y_test)  # y_test가 실제 값이라고 가정
        
        results.append({
            'C': C,
            'fold': i,
            'AUC': auc,
            'IC': ic,
            'p_value': pval
        })

# 결과를 DataFrame으로 변환하여 분석
results_df = pd.DataFrame(results)
```

### AUC와 IC를 사용한 결과 평가

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# AUC와 IC 값 추출
auc_values = results_df.groupby('C')['AUC'].agg(['mean', 'median'])
ic_values = results_df.groupby('C')['IC'].agg(['mean', 'median'])

# AUC 결과 시각화
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(np.log10(Cs), auc_values['median'], marker='o', label='Median AUC')
plt.plot(np.log10(Cs), auc_values['mean'], marker='x', label='Mean AUC')
plt.xlabel('Log10(C)')
plt.ylabel('AUC')
plt.title('AUC by Log10(C)')
plt.legend()

# IC 결과 시각화
plt.subplot(1, 2, 2)
plt.bar(np.log10(Cs), ic_values['mean'], width=0.1, label='Mean IC')
plt.bar(np.log10(Cs) + 0.1, ic_values['median'], width=0.1, label='Median IC')
plt.xlabel('Log10(C)')
plt.ylabel('Information Coefficient (IC)')
plt.title('IC by Log10(C)')
plt.legend()

plt.tight_layout()
plt.show()

# 결과 분석
best_c_mean_auc = Cs[np.argmax(auc_values['mean'])]
best_c_median_auc = Cs[np.argmax(auc_values['median'])]
print(f"Best C for mean AUC: {best_c_mean_auc}")
print(f"Best C for median AUC: {best_c_median_auc}")

# 최적의 C 값에 따른 AUC 및 IC 출력
print("AUC and IC at best C (mean AUC):")
print(auc_values.loc[best_c_mean_auc], ic_values.loc[best_c_mean_auc])
```
